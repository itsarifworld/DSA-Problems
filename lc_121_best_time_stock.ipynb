# 121. Best Time to Buy and Sell Stock

## 🧾 Problem Statement
You are given an array `prices` where `prices[i]` is the price of a stock on the ith day.

You want to maximize your profit by choosing a single day to buy one stock and a different future day to sell it.

- You must buy before you sell.
- If no profit is possible, return 0.


## 💡 Approach (Intuition)

We want to:
- Buy at the lowest price possible
- Sell at the highest price after buying

### Key Idea:
- Track the minimum price seen so far (`mini`)
- At each day, calculate potential profit = current price - minimum price
- Keep updating the maximum profit

This ensures we always buy before selling and get maximum profit.


```python
# Final Optimized Code (O(n))
from typing import List

class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        mini = prices[0]
        profit = 0

        for i in range(len(prices)):
            cost = prices[i] - mini
            profit = max(profit, cost)
            mini = min(mini, prices[i])
        return profit
```

## 🔍 Step-by-Step Dry Run

Let's take:
`prices = [7,1,5,3,6,4]`


```python
prices = [7,1,5,3,6,4]

mini = prices[0]
profit = 0

print("Day | Price | Min So Far | Profit")
print("-----------------------------------")

for i in range(len(prices)):
    cost = prices[i] - mini
    profit = max(profit, cost)
    mini = min(mini, prices[i])
    print(i, "   |", prices[i], "   |", mini, "       |", profit)

print("\nFinal Profit:", profit)
```

## ⚡ Complexity Analysis

- Time Complexity: O(n)
- Space Complexity: O(1)

Only one pass through the array is needed.


## 🚀 Key Takeaways

- Always track minimum price so far
- Calculate profit at each step
- Update maximum profit dynamically
- Greedy approach works perfectly here
