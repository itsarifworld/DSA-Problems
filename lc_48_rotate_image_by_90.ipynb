# 🧩 48. Rotate Image (LeetCode)

## 📌 Problem
Given an `n x n` matrix, rotate it by **90° clockwise in-place**.

👉 You **cannot use extra space** (no new matrix).

---

## 🎯 Approach Overview

We solve this in **2 steps**:
1. **Transpose the matrix** (swap across diagonal)
2. **Reverse each row**

---

## 🎥 Step-by-Step Visualization


````python
import time
import copy

def print_matrix(mat):
    for row in mat:
        print(row)
    print()

matrix = [[1,2,3],[4,5,6],[7,8,9]]

print("Original Matrix:")
print_matrix(matrix)
time.sleep(1)

# Step 1: Transpose
n = len(matrix)
for i in range(n):
    for j in range(i, n):
        matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]

print("After Transpose:")
print_matrix(matrix)
time.sleep(1)

# Step 2: Reverse each row
for row in matrix:
    row.reverse()

print("After Row Reverse (Final Rotated Matrix):")
print_matrix(matrix)
````

## 🚀 Final Code (Optimized)


````python
from typing import List

class Solution:
    def rotate(self, matrix: List[List[int]]) -> None:
        n = len(matrix)

        # Transpose
        for i in range(n):
            for j in range(i, n):
                matrix[i][j], matrix[j][i] = matrix[j][i], matrix[i][j]

        # Reverse each row
        for i in range(n):
            p1, p2 = 0, n - 1
            while p1 < p2:
                matrix[i][p1], matrix[i][p2] = matrix[i][p2], matrix[i][p1]
                p1 += 1
                p2 -= 1
````

## ⏱️ Complexity Analysis

### Time Complexity
- Transpose → `O(n²)`
- Reverse → `O(n²)`
- **Overall → `O(n²)`**

### Space Complexity
- No extra space used
- **O(1)** (in-place)
