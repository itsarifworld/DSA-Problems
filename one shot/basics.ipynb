```python
# Problem 1 — lists + loops
# Given a list of numbers, return a new list
# containing only the unique elements, preserving order.
# Input: [3, 1, 4, 1, 5, 9, 2, 6, 5, 3]
# Output: [3, 1, 4, 5, 9, 2, 6]

def unique_order(arr):
    seen = set()
    result = []

    for num in arr:
        if num not in seen:
            seen.add(num)
            result.append(num)
    return result

```

```python
# Problem 3 — two pointers + while loop
# Check if a string is a palindrome ignoring non-alphanumeric.
# Input: "A man, a plan, a canal: Panama"  Output: True

def is_palindrome(s):
    
    new = "".join(ch for ch in s if ch.isalpha())
    # TWO POINTERS
    lo = 0
    hi = len(new) - 1

    while(lo<= hi):
        if new[lo] != new[hi]:
            return False
        lo += 1
        hi -= 1
    return True

# TIME COMPLEXITY - O(n), Space Complexity - O(n)

def is_palindrome(s):
    s = s.lower()
    left, right = 0, len(s) - 1
    while left < right:
        while left < right and not s[left].isalnum():
            left += 1
        while left < right and not s[right].isalnum():
            right -= 1
        if s[left] != s[right]: return False
        left += 1; right -= 1
    return True
# Time: O(n)  Space: O(1)
```

```python
# Problem 4 — recursion
# Count the number of ways to climb n stairs
# (1 or 2 steps at a time). This is exactly Fibonacci.
# Input: n=4   Output: 5

def climb_stairs(n, memo={}):
    if n <= 2: return n
    if n in memo: return memo[n]
    memo[n] = climb_stairs(n-1, memo) + climb_stairs(n-2, memo)
    return memo[n]
# Time: O(n)  Space: O(n)
```
