
# 🧠 LeetCode 287: Find the Duplicate Number

## 📌 Problem Statement

Given an array of integers `nums` containing **n + 1 integers** where each integer is in the range **[1, n] inclusive**.

There is only **one repeated number**, return this repeated number.

### ⚠️ Constraints:
- You must NOT modify the array
- Use only **constant extra space**

---

## 🧪 Examples

**Example 1**
```
Input: nums = [1,3,4,2,2]
Output: 2
```

**Example 2**
```
Input: nums = [3,1,3,4,2]
Output: 3
```

**Example 3**
```
Input: nums = [3,3,3,3,3]
Output: 3
```



# 🔢 Approach 1: Frequency Array (Hashing)

## 💡 Idea
- Count occurrences using an auxiliary array
- As soon as a number appears twice → return it

## ⚠️ Drawback
- Uses extra space → ❌ violates constraint


````

class Solution:
    def findDuplicate(self, nums):
        freq = [0] * (len(nums) + 1)

        for num in nums:
            freq[num] += 1
            if freq[num] == 2:
                return num
````


### ⏱ Complexity
- Time: O(n)
- Space: O(n)

---

## 🧠 Why it works?
We directly count occurrences. The first number reaching frequency 2 is the duplicate.



# 🔁 Approach 2: Sorting

## 💡 Idea
- Sort the array
- Duplicate elements will become adjacent


````

class Solution:
    def findDuplicate(self, nums):
        nums.sort()
        n = len(nums)

        for i in range(n - 1):
            if nums[i] == nums[i + 1]:
                return nums[i]
````


### ⏱ Complexity
- Time: O(n log n)
- Space: O(1) (ignoring sort space)

---

## ⚠️ Problem
- Modifies array → ❌ violates constraint



# 🐢🐇 Approach 3: Floyd’s Cycle Detection (Optimal)

## 💡 Idea
Treat array as a linked list:
- Index = node
- Value = next pointer

Cycle exists due to duplicate.

---

## 🔁 Step 1: Detect Cycle
Move:
- slow → 1 step
- fast → 2 steps

They meet inside cycle.

## 🎯 Step 2: Find Entrance
Reset fast to start.
Move both 1 step → meeting point = duplicate.


````

class Solution:
    def findDuplicate(self, nums):
        slow = nums[0]
        fast = nums[0]

        # Step 1: detect cycle
        while True:
            slow = nums[slow]
            fast = nums[nums[fast]]
            if slow == fast:
                break

        # Step 2: find entrance
        fast = nums[0]
        while slow != fast:
            slow = nums[slow]
            fast = nums[fast]

        return slow
````


### ⏱ Complexity
- Time: O(n)
- Space: O(1) ✅

---

## 🧠 Why it works?
This uses **Floyd’s Cycle Detection Algorithm**:
- Duplicate creates a cycle
- Meeting point helps locate cycle start

---

## 🏆 Final Verdict

| Approach | Time | Space | Valid? |
|----------|------|------|--------|
| Frequency | O(n) | O(n) | ❌ |
| Sorting | O(n log n) | O(1) | ❌ |
| Floyd’s | O(n) | O(1) | ✅ BEST |

---

## 🚀 Key Takeaway
Whenever:
- Values are in range [1, n]
- Duplicate exists

👉 Think **cycle detection**
