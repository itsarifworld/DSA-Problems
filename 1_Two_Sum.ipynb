# LeetCode Problem 1: Two Sum

**Difficulty**: Easy

## Problem Statement

Given an array of integers `nums` and an integer `target`, return *indices of the two numbers such that they add up to* `target`.

You may assume that each input would have **exactly one solution**, and you may not use the same element twice.

You can return the answer in any order.

### Examples:

**Example 1:**
```
Input: nums = [2, 7, 11, 15], target = 9
Output: [0, 1]
Explanation: nums[0] + nums[1] == 2 + 7 == 9, so we return [0, 1].
```

**Example 2:**
```
Input: nums = [3, 2, 4], target = 6
Output: [1, 2]
```

**Example 3:**
```
Input: nums = [3, 3], target = 6
Output: [0, 1]
```

## Constraints

- 2 <= nums.length <= 10^4
- -10^9 <= nums[i] <= 10^9
- -10^9 <= target <= 10^9
- Only one valid answer exists.


## Approach: Hash Map (Dictionary)

### Intuition
Instead of using a brute force approach (nested loops), we can use a hash map to store the numbers we've seen along with their indices. For each number, we check if its complement (target - number) has already been stored.

### Algorithm
1. Create an empty dictionary to store `{number: index}`
2. Iterate through the array:
   - Calculate the complement: `complement = target - current_number`
   - If the complement exists in the dictionary, return the stored index and current index
   - Otherwise, add the current number and its index to the dictionary

### Complexity Analysis
- **Time Complexity**: O(n) - We iterate through the array once
- **Space Complexity**: O(n) - We store up to n numbers in the hash map


## Solution

````python
def twoSum(nums, target):
    """
    Find indices of two numbers that add up to target.
    
    Args:
        nums: List of integers
        target: Target sum
    
    Returns:
        List of two indices [i, j] where nums[i] + nums[j] == target
    """
    num_to_index = {}  # Dictionary to store {number: index}
    
    for i, num in enumerate(nums):
        complement = target - num  # What we're looking for
        
        # If complement exists in dictionary, we found our pair
        if complement in num_to_index:
            return [num_to_index[complement], i]
        
        # Store current number and its index
        num_to_index[num] = i
    
    # Should never reach here given problem constraints
    return []
````

## Key Takeaways

1. **Hash Map is Key**: Using a dictionary allows us to solve this in O(n) time instead of O(n²) with brute force.
2. **Complement Strategy**: Always ask "what value do I need to find?" and check if it's been seen.
3. **Single Pass**: We only need to iterate through the array once.
4. **Space-Time Trade-off**: We use O(n) extra space to achieve O(n) time complexity.

This is a classic problem that demonstrates the power of hash-based solutions in competitive programming!
