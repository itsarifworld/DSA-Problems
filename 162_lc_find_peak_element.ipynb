# 162_lc_find_peak_element

## Problem Statement
A peak element is an element that is strictly greater than its neighbors.

Given an integer array nums, find a peak element and return its index.

- If the array contains multiple peaks, return any one of them.
- You may imagine that nums[-1] = nums[n] = -∞.
- Time complexity must be O(log n).


## Brute Force Approach

### Idea:
Check every element and verify if it is greater than its neighbors.

### Code:

```python
def findPeakElement_brute(nums):
    n = len(nums)
    for i in range(n):
        left = nums[i-1] if i-1 >= 0 else float('-inf')
        right = nums[i+1] if i+1 < n else float('-inf')
        if nums[i] > left and nums[i] > right:
            return i
```

### Time Complexity:
O(n)


## Better Approach

### Idea:
Try to identify increasing/decreasing pattern but still may scan large portion.
(Still effectively O(n) in worst case)


```python
def findPeakElement_better(nums):
    for i in range(1, len(nums)):
        if nums[i] < nums[i-1]:
            return i-1
    return len(nums) - 1
```

### Time Complexity:
O(n)


## Optimal Approach (Binary Search)

### Idea:
- If nums[mid] < nums[mid+1], peak lies on right
- Else peak lies on left (including mid)

### Code:

```python
def findPeakElement_optimal(nums):
    low, high = 0, len(nums) - 1
    
    while low < high:
        mid = (low + high) // 2
        if nums[mid] < nums[mid + 1]:
            low = mid + 1
        else:
            high = mid
    
    return low
```

### Time Complexity:
O(log n)
