# LeetCode 78 - Subsets

**Difficulty:** Medium  
**Topics:** Array, Backtracking, Bit Manipulation  
**Pattern:** For every element, decide whether to include it or skip it.

## Problem Statement

Given an integer array `nums` of unique elements, return all possible subsets, also called the power set.

The solution set must not contain duplicate subsets. The subsets can be returned in any order.

### Examples

```text
nums = [1, 2, 3]
Output: [[], [1], [2], [1, 2], [3], [1, 3], [2, 3], [1, 2, 3]]
```

```text
nums = [0]
Output: [[], [0]]
```

## Core Idea

A subset is created by making one binary decision for each element:

- Include the element.
- Exclude the element.

For `n` elements, there are `2^n` possible combinations of these decisions. That means any correct algorithm must produce `2^n` subsets, so the output itself is exponential.

## Approach 1: Iterative Expansion

### Idea

Start with the empty subset. For each number, duplicate all existing subsets and add the current number to each duplicate.

For `nums = [1, 2, 3]`:

```text
start: [[]]
add 1: [[], [1]]
add 2: [[], [1], [2], [1, 2]]
add 3: [[], [1], [2], [1, 2], [3], [1, 3], [2, 3], [1, 2, 3]]
```

### Complexity

- Time: `O(n * 2^n)` because we copy up to `n` values for each subset.
- Space: `O(n * 2^n)` for the output.

````python
from typing import List


class IterativeSolution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        result = [[]]

        for num in nums:
            new_subsets = []
            for subset in result:
                new_subsets.append(subset + [num])
            result.extend(new_subsets)

        return result
````

## Approach 2: Bitmasking

### Idea

Every subset can be represented by a binary mask of length `n`.

For `nums = [1, 2, 3]`, the mask `101` means include index `0`, skip index `1`, include index `2`, so the subset is `[1, 3]`.

### Steps

1. Iterate `mask` from `0` to `2^n - 1`.
2. For each bit position `i`, check whether bit `i` is set.
3. If it is set, include `nums[i]` in the current subset.

### Complexity

- Time: `O(n * 2^n)`
- Space: `O(n * 2^n)` for the output.

````python
class BitmaskSolution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        n = len(nums)
        result = []

        for mask in range(1 << n):
            subset = []
            for i in range(n):
                if mask & (1 << i):
                    subset.append(nums[i])
            result.append(subset)

        return result
````

## Approach 3: Backtracking

### Idea

Build subsets using a decision tree. At each index, decide whether to include the current number or skip it.

A common implementation appends the current path as a valid subset, then tries adding each later number.

### Why This Is Useful

Backtracking is the most reusable pattern here. The same structure can be adapted to combinations, permutations, subset sum, and many search problems.

### Complexity

- Time: `O(n * 2^n)`
- Space: `O(n)` recursion depth, excluding the output.

````python
class Solution:
    def subsets(self, nums: List[int]) -> List[List[int]]:
        result = []
        path = []

        def backtrack(start: int) -> None:
            result.append(path.copy())

            for i in range(start, len(nums)):
                path.append(nums[i])
                backtrack(i + 1)
                path.pop()

        backtrack(0)
        return result
````

## Backtracking Walkthrough

For `nums = [1, 2, 3]`:

```text
path = []       -> add []
path = [1]      -> add [1]
path = [1, 2]   -> add [1, 2]
path = [1, 2, 3]-> add [1, 2, 3]
backtrack to [1]
path = [1, 3]   -> add [1, 3]
backtrack to []
path = [2]      -> add [2]
path = [2, 3]   -> add [2, 3]
backtrack to []
path = [3]      -> add [3]
```

The exact order may differ from other accepted solutions, but the set of subsets is the same.

````python
def canonical(subsets: List[List[int]]) -> List[tuple[int, ...]]:
    return sorted(tuple(sorted(subset)) for subset in subsets)


solver = Solution()

assert canonical(solver.subsets([1, 2, 3])) == canonical([
    [], [1], [2], [3], [1, 2], [1, 3], [2, 3], [1, 2, 3]
])
assert canonical(solver.subsets([0])) == canonical([[], [0]])
assert canonical(solver.subsets([])) == canonical([[]])
assert len(solver.subsets([4, 5, 6, 7])) == 16

print("All test cases passed.")
````

## Takeaways

- Subsets are built from include/exclude decisions.
- There are `2^n` subsets, so exponential output size is unavoidable.
- Backtracking is the most flexible solution pattern for this problem.
- Use `path.copy()` when appending to the result so later mutations do not change already saved subsets.
