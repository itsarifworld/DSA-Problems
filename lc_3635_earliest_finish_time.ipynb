# LeetCode 3635 — Earliest Finish Time for Land and Water Rides II

## Idea

Let:

- `land_finish = landStartTime[i] + landDuration[i]`
- `water_finish = waterStartTime[j] + waterDuration[j]`

If we do **land → water**, only the earliest possible land finish time matters.

If we do **water → land**, only the earliest possible water finish time matters.

This leads to an **O(n + m)** solution.


```
from typing import List

class Solution:
    def earliestFinishTime(
        self,
        landStartTime: List[int],
        landDuration: List[int],
        waterStartTime: List[int],
        waterDuration: List[int]
    ) -> int:

        min_land_finish = min(
            s + d for s, d in zip(landStartTime, landDuration)
        )

        min_water_finish = min(
            s + d for s, d in zip(waterStartTime, waterDuration)
        )

        ans = float('inf')

        # land -> water
        for ws, wd in zip(waterStartTime, waterDuration):
            ans = min(ans, wd + max(ws, min_land_finish))

        # water -> land
        for ls, ld in zip(landStartTime, landDuration):
            ans = min(ans, ld + max(ls, min_water_finish))

        return ans
```
