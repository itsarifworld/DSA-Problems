# LeetCode 347 - Top K Frequent Elements

**Difficulty:** Medium  
**Topics:** Array, Hash Table, Heap, Bucket Sort  
**Pattern:** Count frequencies, then keep the highest-frequency values.

## Problem Statement

Given an integer array `nums` and an integer `k`, return the `k` values that appear most often in the array.

The answer can be returned in any order. The input is designed so that the top `k` frequent values are uniquely determined.

### Examples

```text
nums = [1, 1, 1, 2, 2, 3], k = 2
Output: [1, 2]
```

```text
nums = [1], k = 1
Output: [1]
```

## Core Idea

The problem is not asking for the largest numbers. It is asking for the numbers with the largest **frequency**.

So the work naturally splits into two parts:

1. Count how often every value appears.
2. Select the `k` values with the biggest counts.

A hash map is the right tool for counting. After that, we can either sort by frequency or use bucket sort to avoid sorting altogether.

## Approach 1: Frequency Map + Sorting

### Steps

1. Use `Counter` to build a map of `value -> frequency`.
2. Sort the unique values by frequency in descending order.
3. Return the first `k` values.

### Complexity

- Time: `O(n log n)` in the worst case because we may sort up to `n` unique values.
- Space: `O(n)` for the frequency map and sorted unique values.

This approach is simple and reliable, but LeetCode expects better than full sorting for this problem.

````python
from collections import Counter
from typing import List


class SortingSolution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        frequency = Counter(nums)
        values = sorted(frequency.keys(), key=lambda value: frequency[value], reverse=True)
        return values[:k]
````

## Approach 2: Frequency Map + Bucket Sort

### Key Observation

A number cannot appear more than `n` times, where `n = len(nums)`. That means every frequency must be in the range `1` to `n`.

Instead of sorting values by frequency, create buckets where:

```text
bucket[f] = all values that appear exactly f times
```

Then scan the buckets from highest frequency to lowest frequency and collect values until we have `k` answers.

### Why This Is Efficient

We never compare values with each other the way a sorting algorithm does. Counting takes linear time, placing values into buckets takes linear time, and scanning the buckets also takes linear time.

### Complexity

- Time: `O(n)`
- Space: `O(n)`

This is the optimal approach for the usual LeetCode follow-up requirement.

````python
class Solution:
    def topKFrequent(self, nums: List[int], k: int) -> List[int]:
        frequency = Counter(nums)
        buckets = [[] for _ in range(len(nums) + 1)]

        for value, count in frequency.items():
            buckets[count].append(value)

        answer = []
        for count in range(len(buckets) - 1, 0, -1):
            for value in buckets[count]:
                answer.append(value)
                if len(answer) == k:
                    return answer

        return answer
````

## Walkthrough

For `nums = [1, 1, 1, 2, 2, 3]` and `k = 2`:

```text
frequency = {
  1: 3,
  2: 2,
  3: 1
}
```

The buckets look like this:

```text
bucket[3] = [1]
bucket[2] = [2]
bucket[1] = [3]
```

Scanning from the back gives `1` first, then `2`, so the result is `[1, 2]`.

````python
def normalize(values):
    return sorted(values)


solver = Solution()

assert normalize(solver.topKFrequent([1, 1, 1, 2, 2, 3], 2)) == [1, 2]
assert normalize(solver.topKFrequent([1], 1)) == [1]
assert normalize(solver.topKFrequent([4, 4, 4, 6, 6, 7, 7, 7, 7], 2)) == [4, 7]
assert normalize(solver.topKFrequent([-1, -1, -2, -2, -2, 3], 1)) == [-2]

print("All test cases passed.")
````

## Takeaways

- Use a hash map when a problem depends on frequency.
- Sorting by frequency is often the easiest valid solution.
- Bucket sort is useful when the sorting key has a small bounded range.
- Here, frequencies are naturally bounded by `len(nums)`, which lets us solve the problem in linear time.
