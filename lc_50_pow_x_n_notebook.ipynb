# Pow(x, n) - LeetCode Problem

Implement a function `pow(x, n)` that calculates x raised to the power n.

### Examples:
- Input: x = 2.0, n = 10 → Output: 1024.0
- Input: x = 2.1, n = 3 → Output: 9.261
- Input: x = 2.0, n = -2 → Output: 0.25

## Approach: Binary Exponentiation

Instead of multiplying x repeatedly (O(n)), we reduce the problem using:

- If n is even → x^n = (x²)^(n/2)
- If n is odd → x^n = x * x^(n-1)

This reduces time complexity to O(log n).

```
class Solution:
    def myPow(self, x: float, n: int) -> float:
        ans = 1
        neg = n < 0

        if neg:
            n = -n

        while n > 0:
            if n % 2 == 1:
                ans *= x
                n -= 1
            else:
                x *= x
                n //= 2

        if neg:
            ans = 1 / ans

        return ans
```

## Complexity

- Time Complexity: O(log n)
- Space Complexity: O(1)
