# 📓 LC 169: Majority Element
Clean, intuitive, and exam-ready breakdown

## 🟦 Problem Statement
Given an array `nums` of size `n`, return the majority element.

The majority element is the element that appears more than ⌊n / 2⌋ times.

### Examples:
- Input: [3,2,3] → Output: 3  
- Input: [2,2,1,1,1,2,2] → Output: 2  

Constraint: Majority element always exists.


## 🟦 Intuition
- Majority element dominates the array (> n/2)
- Two ways:
  1. Count frequency (hashing)
  2. Cancel out elements (Moore’s Voting)


## 🟦 Solution 1: Hashing (Step-by-Step)

```
nums = [2,2,1,1,1,2,2]

freq = {}

for num in nums:
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1

    print(f"Processing {num} -> {freq}")

for key in freq:
    if freq[key] > len(nums)//2:
        print("Majority Element:", key)
```

### ✅ Final Code (Hashing)

```
from typing import List

class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        freq = {}
        for num in nums:
            if num in freq:
                freq[num] += 1
            else:
                freq[num] = 1

        for i in freq:
            if freq[i] > len(nums)//2:
                return i
```

## 🟦 Solution 2: Moore’s Voting Algorithm (Step-by-Step)

```
nums = [2,2,1,1,1,2,2]

cnt = 0

for num in nums:
    if cnt == 0:
        el = num
        cnt = 1
    elif num == el:
        cnt += 1
    else:
        cnt -= 1

    print(f"num={num}, candidate={el}, count={cnt}")

count_check = nums.count(el)
print("Final Candidate:", el)
print("Count:", count_check)
```

### ✅ Final Code (Moore’s Voting)

```
from typing import List

class Solution:
    def majorityElement(self, nums: List[int]) -> int:
        cnt = 0

        for num in nums:
            if cnt == 0:
                el = num
                cnt = 1
            elif num == el:
                cnt += 1
            else:
                cnt -= 1

        cnt1 = 0
        for num in nums:
            if num == el:
                cnt1 += 1

        if cnt1 > len(nums)//2:
            return el
        return -1
```

## 🟦 Time & Space Complexity

### Hashing:
- Time: O(n)
- Space: O(n)

### Moore’s Voting:
- Time: O(n)
- Space: O(1)
