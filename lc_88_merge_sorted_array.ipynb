# 📘 Merge Sorted Array (LeetCode 88) — Complete Notes

This notebook covers **all approaches** from basic → optimal → interview advanced.


## 🧾 Problem Summary
- Two sorted arrays
- Merge into nums1 **in-place**
- nums1 has extra space


## 🔥 Optimal Approach (LeetCode Best)
👉 Use **3 pointers from the back** to avoid overwriting.


```python
class Solution:
    def merge(self, nums1, m, nums2, n):
        p1 = m - 1
        p2 = n - 1
        i = m + n - 1

        while p1 >= 0 and p2 >= 0:
            if nums1[p1] > nums2[p2]:
                nums1[i] = nums1[p1]
                p1 -= 1
            else:
                nums1[i] = nums2[p2]
                p2 -= 1
            i -= 1

        while p2 >= 0:
            nums1[i] = nums2[p2]
            p2 -= 1
            i -= 1
```

### 🧠 Intuition
- Fill from end
- Avoid overwriting


## 🔍 Dry Run
nums1 = [1,2,3,0,0,0]
nums2 = [2,5,6]

Final → [1,2,2,3,5,6]


## 1️⃣ Brute Force (Extra Space)


```python
def merge_brute(nums1, nums2):
    i = j = 0
    merged = []

    while i < len(nums1) and j < len(nums2):
        if nums1[i] <= nums2[j]:
            merged.append(nums1[i]); i += 1
        else:
            merged.append(nums2[j]); j += 1

    merged.extend(nums1[i:])
    merged.extend(nums2[j:])
    return merged
```

### Complexity
Time: O(n+m)
Space: O(n+m)


## 2️⃣ Better Approach (Sort)


```python
nums1[m:] = nums2
nums1.sort()
```

### Complexity
Time: O((n+m) log(n+m))
Space: O(1)


## 3️⃣ 🔥 Gap Method (Advanced)
Inspired by **Shell Sort**


```python
def nextGap(gap):
    if gap <= 1:
        return 0
    return (gap // 2) + (gap % 2)

def merge_gap(nums1, nums2):
    n, m = len(nums1), len(nums2)
    gap = nextGap(n + m)

    while gap > 0:
        i = 0
        j = gap

        while j < (n + m):
            if j < n and nums1[i] > nums1[j]:
                nums1[i], nums1[j] = nums1[j], nums1[i]

            elif i < n and j >= n and nums1[i] > nums2[j-n]:
                nums1[i], nums2[j-n] = nums2[j-n], nums1[i]

            elif i >= n and nums2[i-n] > nums2[j-n]:
                nums2[i-n], nums2[j-n] = nums2[j-n], nums2[i-n]

            i += 1
            j += 1

        gap = nextGap(gap)
```

### 🧠 Intuition
- Compare distant elements
- Gradually reduce gap


### Complexity
Time: O((n+m) log(n+m))
Space: O(1)
