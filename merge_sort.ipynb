# 🔀 Merge Sort
---
> **Category:** Divide and Conquer | Sorting  
> **Difficulty:** Medium  
> **Language:** Python

## 📋 Problem Statement

Given an unsorted array of `n` integers, sort the array in **ascending order** using the **Merge Sort** algorithm.

### Input
- An array `arr` of `n` integers (can be positive, negative, or zero)

### Output
- The same array sorted in non-decreasing (ascending) order

### Examples

| Input | Output |
|-------|--------|
| `[38, 27, 43, 3, 9, 82, 10]` | `[3, 9, 10, 27, 38, 43, 82]` |
| `[5, 1, 4, 2, 8]` | `[1, 2, 4, 5, 8]` |
| `[1]` | `[1]` |
| `[]` | `[]` |

### Constraints
- `0 <= n <= 10^5`
- `-10^9 <= arr[i] <= 10^9`

## 💡 Intuition & Algorithm Design

Merge Sort follows the classic **Divide and Conquer** paradigm:

```
┌─────────────────────────────────────┐
│           DIVIDE AND CONQUER        │
│                                     │
│  1. DIVIDE  → Split array in half   │
│  2. CONQUER → Recursively sort each │
│  3. COMBINE → Merge two sorted halves│
└─────────────────────────────────────┘
```

### Key Insight
Merging two **already-sorted** arrays into one sorted array is an O(n) operation.  
By recursively breaking the problem down to single-element arrays (trivially sorted), we can build back up via efficient merging.

### Two Core Functions
1. **`merge_sort(arr)`** — Recursively splits and calls itself on halves
2. **`merge(left, right)`** — Merges two sorted arrays in O(n) time

### Phase 1: DIVIDE (Splitting)

```
Level 0 (original):         [38, 27, 43, 3]
                              /            \
Level 1 (split in half): [38, 27]       [43, 3]
                           /    \         /    \
Level 2 (single elems): [38]  [27]     [43]   [3]
```

---

### Phase 2: MERGE (Combining — bottom up)

**Step 1: Merge [38] and [27]**
```
left  = [38]    pointer i=0
right = [27]    pointer j=0
result = []

  Compare left[0]=38 vs right[0]=27
  → 27 < 38, so append 27   → result = [27],  j=1 (right exhausted)
  → Append remaining left   → result = [27, 38]

✅ Merged: [27, 38]
```

**Step 2: Merge [43] and [3]**
```
left  = [43]    pointer i=0
right = [3]     pointer j=0
result = []

  Compare left[0]=43 vs right[0]=3
  → 3 < 43, so append 3    → result = [3],  j=1 (right exhausted)
  → Append remaining left  → result = [3, 43]

✅ Merged: [3, 43]
```

**Step 3: Merge [27, 38] and [3, 43]**
```
left  = [27, 38]    pointer i=0
right = [3,  43]    pointer j=0
result = []

  Compare left[0]=27 vs right[0]=3
  → 3 < 27,  append 3    → result = [3],       i=0, j=1

  Compare left[0]=27 vs right[1]=43
  → 27 < 43, append 27   → result = [3, 27],   i=1, j=1

  Compare left[1]=38 vs right[1]=43
  → 38 < 43, append 38   → result = [3, 27, 38], i=2 (left exhausted)

  → Append remaining right → result = [3, 27, 38, 43]

✅ Final Merged: [3, 27, 38, 43]
```

## ⏱️ Time Complexity & 🧠 Space Complexity

### Time Complexity

| Case | Complexity | Reason |
|------|------------|--------|
| **Best Case** | O(n log n) | Even on sorted input, we still split and merge |
| **Average Case** | O(n log n) | T(n) = 2T(n/2) + O(n) by Master Theorem |
| **Worst Case** | O(n log n) | No degenerate case exists unlike QuickSort |

**Recurrence Relation:**
```
T(n) = 2·T(n/2) + O(n)
     ↓ Master Theorem (Case 2: a=2, b=2, f(n)=n, n^log_b(a) = n^1)
T(n) = O(n log n)
```

**Why log n levels?**  
Each split halves the array → we need log₂(n) levels to reach size-1 subarrays.  
Each level does O(n) total work across all merges → **O(n log n)** overall.

```
Level 0: 1 array  of size n     →  n   comparisons
Level 1: 2 arrays of size n/2   →  n   comparisons
Level 2: 4 arrays of size n/4   →  n   comparisons
  ...                               ...
Level k: n arrays of size 1     →  n   comparisons
         ← log₂(n) levels →    = n·log n  total
```

---

### Space Complexity

| Component | Space | Reason |
|-----------|-------|--------|
| **Auxiliary arrays** | O(n) | Temporary left/right arrays during merge |
| **Call stack** | O(log n) | Recursion depth = log₂(n) |
| **Total** | **O(n)** | Dominated by auxiliary space |

> ⚠️ Merge Sort is **NOT in-place**. It requires O(n) extra space, which is its main disadvantage compared to in-place sorts like QuickSort or HeapSort.

---

### Comparison with Other Sorts

| Algorithm | Best | Average | Worst | Space | Stable |
|-----------|------|---------|-------|-------|--------|
| **Merge Sort** | O(n log n) | O(n log n) | O(n log n) | O(n) | ✅ Yes |
| Quick Sort | O(n log n) | O(n log n) | O(n²) | O(log n) | ❌ No |
| Heap Sort | O(n log n) | O(n log n) | O(n log n) | O(1) | ❌ No |
| Bubble Sort | O(n) | O(n²) | O(n²) | O(1) | ✅ Yes |
| Insertion Sort | O(n) | O(n²) | O(n²) | O(1) | ✅ Yes |

## 💻 Implementation

````python
def merge_sort(arr):
    """
    Sorts an array using the Merge Sort algorithm.

    Args:
        arr (list): Unsorted list of comparable elements.

    Returns:
        list: A new sorted list in ascending order.

    Time Complexity:  O(n log n) — all cases
    Space Complexity: O(n)      — auxiliary arrays + O(log n) call stack
    """

    # ── BASE CASE ──────────────────────────────────────────────────
    # An array of 0 or 1 element is already sorted
    if len(arr) <= 1:
        return arr

    # ── DIVIDE ─────────────────────────────────────────────────────
    # Find the midpoint and split the array into two halves
    mid = len(arr) // 2
    left_half  = arr[:mid]   # Elements from index 0   to mid-1
    right_half = arr[mid:]   # Elements from index mid to end

    # ── CONQUER ────────────────────────────────────────────────────
    # Recursively sort each half
    left_sorted  = merge_sort(left_half)
    right_sorted = merge_sort(right_half)

    # ── COMBINE ────────────────────────────────────────────────────
    # Merge the two sorted halves
    return merge(left_sorted, right_sorted)


def merge(left, right):
    """
    Merges two sorted arrays into a single sorted array.

    Args:
        left  (list): Sorted left half.
        right (list): Sorted right half.

    Returns:
        list: Merged sorted list.

    Time Complexity:  O(n)  where n = len(left) + len(right)
    Space Complexity: O(n)  for the result array
    """
    result = []
    i = 0  # Pointer for left array
    j = 0  # Pointer for right array

    # Compare elements from both halves and pick the smaller one
    while i < len(left) and j < len(right):
        if left[i] <= right[j]:   # <= ensures stability
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1

    # Append any remaining elements from left or right
    result.extend(left[i:])   # left elements not yet picked
    result.extend(right[j:])  # right elements not yet picked

    return result
````

## ✅ Test Cases

````python
# ─── Test Helper ──────────────────────────────────────────────────
def run_test(description, arr):
    result = merge_sort(arr)
    expected = sorted(arr)
    status = "✅ PASS" if result == expected else "❌ FAIL"
    print(f"{status} | {description}")
    print(f"       Input:    {arr}")
    print(f"       Output:   {result}")
    print(f"       Expected: {expected}")
    print()

# ─── Run Tests ────────────────────────────────────────────────────
run_test("Basic example from dry run",   [38, 27, 43, 3])
run_test("Larger unsorted array",         [38, 27, 43, 3, 9, 82, 10])
run_test("Already sorted",                [1, 2, 3, 4, 5])
run_test("Reverse sorted",                [5, 4, 3, 2, 1])
run_test("All same elements",             [7, 7, 7, 7])
run_test("Single element",                [42])
run_test("Two elements",                  [2, 1])
run_test("Empty array",                   [])
run_test("Negative numbers",              [-3, -1, -7, 0, 2])
run_test("Mixed positive and negative",   [3, -2, 0, -5, 8, 1])
````

## 🔍 Verbose Dry Run (With Print Tracing)

````python
def merge_sort_verbose(arr, depth=0):
    """Merge sort with step-by-step printing to visualize recursion."""
    indent = "  " * depth
    print(f"{indent}merge_sort({arr})")

    if len(arr) <= 1:
        print(f"{indent}  → base case, return {arr}")
        return arr

    mid = len(arr) // 2
    left  = arr[:mid]
    right = arr[mid:]

    print(f"{indent}  → DIVIDE: left={left}, right={right}")

    left_sorted  = merge_sort_verbose(left,  depth + 1)
    right_sorted = merge_sort_verbose(right, depth + 1)

    merged = merge(left_sorted, right_sorted)
    print(f"{indent}  → MERGE({left_sorted}, {right_sorted}) = {merged}")
    return merged


print("=" * 55)
print(" VERBOSE DRY RUN: merge_sort([38, 27, 43, 3])")
print("=" * 55)
result = merge_sort_verbose([38, 27, 43, 3])
print()
print(f"Final Sorted Output: {result}")
````

## 📊 Performance Benchmarking

````python
import random
import time

print(f"{'Input Size':<15} {'Time (seconds)':<20} {'Complexity Check (T/n*log n)'}")
print("-" * 60)

import math

sizes = [100, 1_000, 10_000, 100_000]

for n in sizes:
    data = [random.randint(-10**6, 10**6) for _ in range(n)]
    start = time.time()
    merge_sort(data)
    elapsed = time.time() - start

    # Normalise by n*log2(n) to confirm O(n log n)
    norm = elapsed / (n * math.log2(n)) * 1e8  # scaled for readability
    print(f"{n:<15} {elapsed:<20.6f} {norm:.4f}  (should be ~constant)")
````

## 🌟 Variants & Optimizations

### 1. In-Place Merge Sort
Reduces space to O(log n) but is harder to implement and slower in practice.

### 2. Bottom-Up (Iterative) Merge Sort
Avoids recursion overhead. Starts with size-1 subarrays and iteratively doubles the merge window.

### 3. Natural Merge Sort
Detects already-sorted runs in the input — performs better on partially sorted data, achieving O(n) on already-sorted input.

### 4. Timsort (Python's built-in sort!)
A hybrid of Merge Sort + Insertion Sort, used by Python's `sorted()` and `.sort()`.  
Uses Insertion Sort for small subarrays (< 64 elements) and Natural Merge Sort for the rest.

| **Online** | ❌ No (needs full array) |

### When to use Merge Sort?
- ✅ When **stability** is required (preserving relative order of equal elements)
- ✅ When sorting **linked lists** (no random access needed)
- ✅ When dealing with **external sorting** (data too large for RAM, stored on disk)
- ❌ Avoid when memory is constrained (prefer HeapSort / in-place QuickSort)
