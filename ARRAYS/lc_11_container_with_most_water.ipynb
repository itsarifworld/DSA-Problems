# LeetCode 11: Container With Most Water

## Problem Statement

You are given an integer array `height` of length `n`.

There are `n` vertical lines drawn such that the two endpoints of the `i-th` line are:

- `(i, 0)`
- `(i, height[i])`

Find two lines that together with the x-axis form a container such that the container contains the maximum amount of water.

Return the maximum amount of water a container can store.


# Approach 1: Brute Force

## Idea
Check every pair and calculate area.

## Pseudocode

```text
ans = 0

for i in range(n):
    for j in range(i+1, n):

        width = j - i
        height = min(height[i], height[j])

        area = width * height

        ans = max(ans, area)

return ans
```

## Time Complexity
- O(n²)

## Space Complexity
- O(1)


````
class Solution:
    def maxArea(self, height):

        ans = 0
        n = len(height)

        for i in range(n):
            for j in range(i + 1, n):

                width = j - i
                ht = min(height[i], height[j])

                area = width * ht

                ans = max(ans, area)

        return ans


sol = Solution()
print(sol.maxArea([1,8,6,2,5,4,8,3,7]))
````

# Approach 2: Two Pointer Approach

## Idea

- Start with maximum width
- Move the smaller height pointer
- Try to find a taller wall

## Pseudocode

```text
left = 0
right = n - 1

while left < right:

    width = right - left
    height = min(height[left], height[right])

    area = width * height

    update maximum area

    if height[left] < height[right]:
        left += 1
    else:
        right -= 1
```

## Time Complexity
- O(n)

## Space Complexity
- O(1)


````
class Solution:
    def maxArea(self, height):

        left = 0
        right = len(height) - 1

        maxm = 0

        while left < right:

            width = right - left
            ht = min(height[left], height[right])

            area = width * ht

            maxm = max(maxm, area)

            if height[left] < height[right]:
                left += 1
            else:
                right -= 1

        return maxm


sol = Solution()
print(sol.maxArea([1,8,6,2,5,4,8,3,7]))
````
