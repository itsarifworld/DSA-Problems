# LeetCode 74 — Search a 2D Matrix

## Problem Statement

You are given an `m x n` integer matrix with:

1. Each row sorted in non-decreasing order.
2. The first element of each row is greater than the last element of the previous row.

Given a target value, return `True` if it exists in the matrix, otherwise return `False`.

**Required complexity:** `O(log(m*n))`


# Key Observation

Because:

- Rows are sorted.
- Every row starts after the previous row ends.

The matrix behaves like one single sorted array.

Example:

```
[
 [1, 3, 5, 7],
 [10,11,16,20],
 [23,30,34,60]
]

Equivalent sorted array:

[1,3,5,7,10,11,16,20,23,30,34,60]
```


# Solution 1: Brute Force Scan

## Idea

Check every element one by one until the target is found.

## Approach

1. Iterate through every row.
2. Iterate through every number in the row.
3. Return True if target is found.
4. Otherwise return False.


````
def searchMatrix(matrix, target):
    for row in matrix:
        for num in row:
            if num == target:
                return True
    return False
````

## Complexity Analysis

Let:

- m = number of rows
- n = number of columns

### Time Complexity

In worst case we visit every element:

T(m,n) = m × n

**Time Complexity = O(mn)**

### Space Complexity

Only a few variables are used.

**Space Complexity = O(1)**


# Solution 2: Find Candidate Row + Linear Search

## Idea

Since rows are ordered, first determine which row could contain the target.

Then search only inside that row.

## Approach

1. Iterate through rows.
2. Check whether target lies between first and last element.
3. If yes, search inside that row.
4. Otherwise move to next row.


````
def searchMatrix(matrix, target):
    for row in matrix:
        if row[0] <= target <= row[-1]:
            for num in row:
                if num == target:
                    return True
    return False
````

## Complexity Analysis

### Time Complexity

Finding row: O(m)

Searching row: O(n)

Total:

O(m+n)

### Space Complexity

O(1)


# Solution 3: Binary Search on Row + Binary Search Inside Row

## Idea

1. Use binary search to locate the correct row.
2. Use binary search again inside that row.

This takes advantage of row ordering.


````
def searchMatrix(matrix, target):
    m, n = len(matrix), len(matrix[0])

    top, bottom = 0, m - 1

    while top <= bottom:
        mid = (top + bottom) // 2

        if target < matrix[mid][0]:
            bottom = mid - 1
        elif target > matrix[mid][-1]:
            top = mid + 1
        else:
            row = mid
            break
    else:
        return False

    left, right = 0, n - 1

    while left <= right:
        mid = (left + right) // 2

        if matrix[row][mid] == target:
            return True
        elif matrix[row][mid] < target:
            left = mid + 1
        else:
            right = mid - 1

    return False
````

## Complexity Analysis

### Finding Row

Binary search among m rows:

O(log m)

### Finding Element

Binary search among n columns:

O(log n)

### Total

O(log m + log n)

Using logarithm properties:

O(log(mn))

### Space Complexity

O(1)


# Solution 4 (Optimal): Treat Matrix as a Single Sorted Array

## Core Insight

Imagine flattening the matrix:

```
[1,3,5,7,10,11,16,20,23,30,34,60]
```

Instead of physically creating this array, map indices mathematically.

For a matrix with n columns:

row = mid // n

col = mid % n

This allows binary search directly on the virtual array.

This is the most elegant and interview-preferred solution.


````
def searchMatrix(matrix, target):
    m, n = len(matrix), len(matrix[0])

    left = 0
    right = m * n - 1

    while left <= right:
        mid = (left + right) // 2

        row = mid // n
        col = mid % n

        value = matrix[row][col]

        if value == target:
            return True
        elif value < target:
            left = mid + 1
        else:
            right = mid - 1

    return False
````

## Detailed Complexity Calculation

Let:

- m = rows
- n = columns

Total elements:

m × n

Binary search repeatedly halves the search space.

Number of iterations:

log₂(m × n)

Therefore:

### Time Complexity

O(log(m × n))

### Space Complexity

Only a few variables:

- left
- right
- mid
- row
- col

No extra data structure.

O(1)

This satisfies the problem requirement exactly.
