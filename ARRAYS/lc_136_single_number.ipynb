# LeetCode 136: Single Number
### Difficulty: Easy | Topic: Bit Manipulation, Hash Table

---

## 1. Problem Statement
Given a **non-empty** array of integers `nums`, every element appears *twice* except for one. Find that single one.

You must implement a solution with a **linear runtime complexity** ($O(n)$) and use **only constant extra space** ($O(1)$).

### Examples:
- **Example 1:**
  - **Input:** `nums = [2, 2, 1]`
  - **Output:** `1`
- **Example 2:**
  - **Input:** `nums = [4, 1, 2, 1, 2]`
  - **Output:** `4`
- **Example 3:**
  - **Input:** `nums = [1]`
  - **Output:** `1`

### Constraints:
- $1 \le \text{nums.length} \le 3 \cdot 10^4$
- $-3 \cdot 10^4 \le \text{nums}[i] \le 3 \cdot 10^4$
- Each element in the array appears twice except for one element which appears only once.

## 2. Approach 1: Hash Map (Frequency Tracking)

### Core Concept:
Iterate through the array and store the frequency of each element in a hash map (dictionary). After counting frequencies, iterate through the hash map to find the element that has a frequency of `1`.

### Pseudocode:
```text
FUNCTION singleNumberHash(nums):
    CREATE empty hash map 'freq'
    FOR each num in nums:
        freq[num] = freq.get(num, 0) + 1
    
    FOR each key in freq:
        IF freq[key] == 1:
            RETURN key
```

### Complexity:
- **Time Complexity:** $O(n)$ — We traverse the list of size $n$ to populate the hash map, and then traverse the hash map keys (at most $n/2 + 1$ keys).
- **Space Complexity:** $O(n)$ — In the worst case, the hash map stores approximately $n/2$ unique elements, consuming linear extra space. This *violates* the problem's strict $O(1)$ space constraint but serves as a great starting benchmark.

````python
from typing import List

class SolutionHash:
    def singleNumber(self, nums: List[int]) -> int:
        freq = {}
        for num in nums:
            freq[num] = freq.get(num, 0) + 1
        
        for key in freq:
            if freq[key] == 1:
                return key

# Test Code
sol_hash = SolutionHash()
print("Hash Map Solution Output:", sol_hash.singleNumber([4, 1, 2, 1, 2]))
````

## 3. Approach 2: Bitwise XOR (Optimal Solution)

### Core Concept:
By leveraging bitwise operations, we can meet the strict $O(1)$ space requirement. The Bitwise XOR (`^`) operator satisfies three critical algebraic rules:
1. **Identity:** $X \oplus 0 = X$
2. **Self-Inverse:** $X \oplus X = 0$
3. **Commutative & Associative:** The order of processing operations does not matter ($A \oplus B \oplus A = A \oplus A \oplus B = 0 \oplus B = B$).

When we XOR all elements in the array together, duplicate pairs cancel out perfectly to `0`, leaving behind only the single unique integer.

### Pseudocode:
```text
FUNCTION singleNumberXOR(nums):
    SET res = 0
    FOR each num in nums:
        res = res XOR num
    RETURN res
```

### Complexity:
- **Time Complexity:** $O(n)$ — We loop through the array of length $n$ exactly once.
- **Space Complexity:** $O(1)$ — Only a single integer variable (`res`) is tracked, using completely constant auxiliary memory.

### 4. Detailed Dry Run (XOR Approach)
Let's visualize the operation sequence for the input configuration: **`nums = [4, 1, 2, 1, 2]`**

| Step | Element (`num`) | Bitwise Operation | Running Value of `res` (Decimal) | Running Value of `res` (Binary) |
| :--- | :--- | :--- | :--- | :--- |
| Initial | — | — | `0` | `0000` |
| 1 | `4` | `0 ^ 4` | `4` | `0100` |
| 2 | `1` | `4 ^ 1` | `5` | `0101` |
| 3 | `2` | `5 ^ 2` | `7` | `0111` |
| 4 | `1` | `7 ^ 1` | `6` | `0110` |
| 5 | `2` | `6 ^ 2` | `4` | `0100` |

**Final Return Value:** `4`

````python
class SolutionXOR:
    def singleNumber(self, nums: List[int]) -> int:
        res = 0
        for num in nums:
            res ^= num
        return res

# Test Code
sol_xor = SolutionXOR()
print("Optimal XOR Solution Output:", sol_xor.singleNumber([4, 1, 2, 1, 2]))
````
