# LeetCode 540: Single Element in a Sorted Array

## Problem Statement
You are given a sorted array consisting of only integers where every element appears exactly twice, except for one element which appears exactly once.

Return the single element that appears only once.

Your solution must run in $O(\log n)$ time and $O(1)$ space.

**Example 1:**
- Input: `nums = [1,1,2,3,3,4,4,8,8]`
- Output: `2`

**Example 2:**
- Input: `nums = [3,3,7,7,10,11,11]`
- Output: `10`

## 1. Brute Force Approach
### Approach
Since the array is sorted, we can iterate through the array in steps of 2. In a normal scenario where every element has a pair, `nums[i]` should always equal `nums[i+1]`. The first time this condition fails, `nums[i]` is our single element. If we reach the end of the loop without returning, the last element must be the single one.

### Complexity
- **Time Complexity:** $O(n)$ - We traverse the array once.
- **Space Complexity:** $O(1)$ - No extra space used.

```python
class Solution:
    def singleNonDuplicate(self, nums: list[int]) -> int:
        # Iterate by 2
        for i in range(0, len(nums) - 1, 2):
            if nums[i] != nums[i + 1]:
                return nums[i]
        
        # If not found in pairs, it's the last element
        return nums[-1]
```

## 2. Optimal Approach: Binary Search
### Theoretical Logic
To achieve $O(\log n)$, we use Binary Search. The key observation is the **index parity** of the pairs:
1. **Before the single element:** The first occurrence of a pair is at an **even** index, and the second is at an **odd** index ($E, O$).
2. **After the single element:** The first occurrence is at an **odd** index, and the second is at an **even** index ($O, E$).

### Detailed Step-by-Step Approach
1. **Initialize** `low = 0` and `high = n - 1`.
2. **Boundary Checks:** If `mid` is at the start or end, check if it's the unique element to avoid index out of bounds.
3. **Check Mid:** If `nums[mid]` is not equal to `nums[mid-1]` AND not equal to `nums[mid+1]`, we found the answer.
4. **Determine the Half:**
   - If `mid` is **Even**:
     - If `nums[mid] == nums[mid+1]`, we are still in the "left side" (before the single element). Move `low = mid + 1`.
     - Else, we are in the "right side". Move `high = mid - 1`.
   - If `mid` is **Odd**:
     - If `nums[mid] == nums[mid-1]`, we are in the "left side". Move `low = mid + 1`.
     - Else, we are in the "right side". Move `high = mid - 1`.

### Dry Run
**Input:** `nums = [1, 1, 2, 3, 3, 4, 4]`

1. **Initial:** `low = 0`, `high = 6`
2. **Iteration 1:**
   - `mid = 3` (Value: `3`)
   - `mid` is **Odd**. `nums[3]` is `3`, `nums[2]` is `2`.
   - Since `nums[mid] != nums[mid-1]`, we are in the **right side** of the unique element.
   - `high = mid - 1 = 2`.
3. **Iteration 2:**
   - `low = 0`, `high = 2`
   - `mid = 1` (Value: `1`)
   - `mid` is **Odd**. `nums[1]` is `1`, `nums[0]` is `1`.
   - Since `nums[mid] == nums[mid-1]`, we are in the **left side**.
   - `low = mid + 1 = 2`.
4. **Iteration 3:**
   - `low = 2`, `high = 2`
   - `mid = 2` (Value: `2`)
   - **Check Unique:** `nums[2]` is different from `nums[1]` and `nums[3]`.
   - **Return 2.**

```python
class Solution:
    def singleNonDuplicate(self, nums: list[int]) -> int:
        n = len(nums)
        low = 0
        high = n - 1
        
        if (n == 1):
            return nums[0]
            
        while(low <= high):
            mid = low + ( high - low )//2
            
            # Boundary checks to prevent index errors
            if(mid == 0 and nums[mid] != nums[mid+1]):
                return nums[mid]
            if(mid == n-1 and nums[mid] != nums[mid-1]):
                return nums[mid]
            if (mid > 0 and mid < n-1 and nums[mid-1] != nums[mid] != nums[mid+1]):
                return nums[mid]

            # Check Parity and move pointers
            if (mid % 2 == 0): # EVEN index
                if(nums[mid] == nums[mid+1]):
                    # Pair is (E, O) as expected -> move right
                    low = mid + 1
                else:
                    # Pair is broken -> move left
                    high = mid - 1
            else: # ODD index
                if(nums[mid] == nums[mid-1]):
                    # Pair is (E, O) as expected -> move right
                    low = mid + 1
                else:
                    # Pair is broken -> move left
                    high = mid - 1
                    
        return nums[low]
```
