# LeetCode 53 - Maximum Subarray (Kadane's Algorithm)

This notebook explains Kadane's Algorithm for solving the Maximum Subarray problem efficiently.

## Problem Statement

Given an integer array `nums`, find the contiguous subarray with the largest sum and return its sum.

## Intuition

- At every index, decide whether:
  - Start a new subarray from the current element
  - Or extend the previous subarray

- Keep track of:
  - `current_sum` → best sum ending at current index
  - `max_sum` → best sum found so far

```
def maxSubArray(nums):
    current_sum = nums[0]
    max_sum = nums[0]

    for i in range(1, len(nums)):
        current_sum = max(nums[i], current_sum + nums[i])
        max_sum = max(max_sum, current_sum)

    return max_sum


nums = [-2,1,-3,4,-1,2,1,-5,4]
print("Maximum Subarray Sum:", maxSubArray(nums))
```

## Time Complexity

- Time Complexity: `O(n)`
- Space Complexity: `O(1)`
