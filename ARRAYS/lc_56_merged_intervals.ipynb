# LeetCode 56 - Merge Intervals

## Problem Statement

Given an array of intervals where `intervals[i] = [starti, endi]`, merge all overlapping intervals, and return an array of the non-overlapping intervals that cover all the intervals in the input.

### Example 1
**Input:** `[[1,3],[2,6],[8,10],[15,18]]`

**Output:** `[[1,6],[8,10],[15,18]]`

### Example 2
**Input:** `[[1,4],[4,5]]`

**Output:** `[[1,5]]`

### Constraints
- `1 <= intervals.length <= 10^4`
- `intervals[i].length == 2`
- `0 <= starti <= endi <= 10^4`


# Intuition

Two intervals overlap if the start of the current interval is less than or equal to the end of the previous interval.

Example:

`[1,3]` and `[2,6]` overlap because `2 <= 3`.

To efficiently merge intervals:

1. Sort intervals by starting time.
2. Start with the first interval.
3. Compare every next interval with the last merged interval.
4. If they overlap, extend the ending point.
5. Otherwise, add the interval as a new merged interval.

Sorting ensures that all potential overlapping intervals appear next to each other.


# Optimal Approach (Sorting + Greedy)

### Step 1
Sort intervals according to their start value.

Example:

Before sorting:
```
[[4,7],[1,4]]
```

After sorting:
```
[[1,4],[4,7]]
```

### Step 2
Initialize result with the first interval.

### Step 3
For every remaining interval:

- If current start <= last merged end:
  - Merge them by updating the end value.
- Else:
  - Add the interval to the result.

### Step 4
Return the result.


````
class Solution:
    def merge(self, intervals):
        intervals.sort(key=lambda x: x[0])

        merged = [intervals[0]]

        for start, end in intervals[1:]:
            last_end = merged[-1][1]

            if start <= last_end:
                merged[-1][1] = max(last_end, end)
            else:
                merged.append([start, end])

        return merged
````

# Time Complexity Analysis

### Sorting
Sorting `n` intervals requires:

```
O(n log n)
```

### Traversal
We visit every interval once:

```
O(n)
```

### Total Time Complexity

```
O(n log n)
```

The sorting step dominates the overall complexity.

# Space Complexity Analysis

Ignoring the output array:

```
O(1)
```

Including the result list:

```
O(n)
```

where `n` is the number of intervals.
