# LeetCode 121: Best Time to Buy and Sell Stock

## 1. Problem Statement
You are given an array `prices` where `prices[i]` is the price of a given stock on the $i^{th}$ day.

You want to maximize your profit by choosing a **single day** to buy one stock and choosing a **different day in the future** to sell that stock.

Return *the maximum profit you can achieve from this transaction*. If you cannot achieve any profit, return `0`.

### Examples
**Example 1:**
* **Input:** `prices = [7, 1, 5, 3, 6, 4]`
* **Output:** `5`
* **Explanation:** Buy on day 2 (price = 1) and sell on day 5 (price = 6), profit = `6 - 1 = 5`.

**Example 2:**
* **Input:** `prices = [7, 6, 4, 3, 1]`
* **Output:** `0`
* **Explanation:** No transactions are done, max profit = `0`.

## 2. Approach: One-Pass (Greedy / Dynamic Programming)
To maximize profit, we want to find the largest peak following the lowest valley. We maintain two tracking variables:
1. `min_price`: Tracks the lowest stock price seen so far. Initialize this to `prices[0]`.
2. `max_profit`: Tracks the maximum profit we can make up to the current day. Initialize this to `0`.

As we loop through each day's price:
* If the current price is less than `min_price`, we update our `min_price` because it represents a better buying point.
* Otherwise, if selling at the current price yields a higher profit than our current `max_profit` (`price - min_price > max_profit`), we update `max_profit`.

## 3. Pseudocode
```text
FUNCTION maxProfit(prices):
    IF prices is empty:
        RETURN 0
        
    min_price = prices[0]
    max_profit = 0
    
    FOR EACH price IN prices:
        IF price < min_price:
            min_price = price
        ELSE IF price - min_price > max_profit:
            max_profit = price - min_price
            
    RETURN max_profit
```

## 4. Dry Run
Let's dry run the algorithm with `prices = [7, 1, 5, 3, 6, 4]`.

| Day ($i$) | Price | `min_price` (Before) | Action / Update Condition | `min_price` (After) | `max_profit` |
|:---:|:---:|:---:|:---|:---:|:---:|
| Initial | - | - | Initialize variables | 7 | 0 |
| 0 | 7 | 7 | `7 < 7` (False), `7 - 7 > 0` (False) | 7 | 0 |
| 1 | 1 | 7 | `1 < 7` -> Update `min_price` | 1 | 0 |
| 2 | 5 | 1 | `5 < 1` (False), `5 - 1 = 4 > 0` -> Update `max_profit` | 1 | 4 |
| 3 | 3 | 1 | `3 < 1` (False), `3 - 1 = 2 > 4` (False) | 1 | 4 |
| 4 | 6 | 1 | `6 < 1` (False), `6 - 1 = 5 > 4` -> Update `max_profit` | 1 | 5 |
| 5 | 4 | 1 | `4 < 1` (False), `4 - 1 = 3 > 5` (False) | 1 | 5 |

**Final Output:** `5`

## 5. Python Code Implementation

````python
from typing import List

class Solution:
    def maxProfit(self, prices: List[int]) -> int:
        if not prices:
            return 0
        
        max_profit = 0
        min_price = prices[0]
        
        for price in prices:
            if price < min_price:
                min_price = price
            elif price - min_price > max_profit:
                max_profit = price - min_price
                
        return max_profit

# Test the solution
sol = Solution()
example_1 = [7, 1, 5, 3, 6, 4]
example_2 = [7, 6, 4, 3, 1]

print(f"Max profit for {example_1}: {sol.maxProfit(example_1)}")
print(f"Max profit for {example_2}: {sol.maxProfit(example_2)}")
````

## 6. Complexity Analysis
* **Time Complexity:** $\mathcal{O}(n)$ — We iterate through the list exactly once.
* **Space Complexity:** $\mathcal{O}(1)$ — We use a constant amount of auxiliary memory.
