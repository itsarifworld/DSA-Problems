# LeetCode 31 — Next Permutation

## Problem Information

- Problem Number: 31
- Difficulty: Medium
- Topics: Array, Two Pointers
- Core Pattern: Lexicographical Permutation

## Problem Statement

Implement the next lexicographically greater permutation in-place.
If no such permutation exists, rearrange into the lowest possible order.


## Pattern Recognition

### Keywords

- Next greater arrangement
- Lexicographical order
- In-place modification
- Constant extra space

### Pattern

Array + Greedy + Two Pointers


## Brute Force Solution

### Thought Process

Generate all permutations, sort them, locate the current permutation, and return the next one.

### Why It Works

Every possible arrangement is examined.

### Why It Is Bad

The number of permutations grows as n!.


## Better / Recursive Solution

Use backtracking to generate permutations manually.

Educational for recursion, but still factorial complexity.


## Optimal Solution

### Key Observation

The suffix of the array is often already in descending order.

We only need to make the smallest possible increase from the right side.

### Steps

1. Find pivot.
2. Find rightmost larger element.
3. Swap.
4. Reverse suffix.


## Detailed Dry Run

Input:

[1,3,5,4,2]

Pivot = 3

Swap with 4

[1,4,5,3,2]

Reverse suffix:

[1,4,2,3,5]


## Correctness Proof

Claim 1:
Pivot is the first position that can be increased.

Claim 2:
Choosing the rightmost larger element creates the minimum increase.

Claim 3:
Reversing the suffix produces the smallest possible suffix.

Therefore the resulting permutation is the next lexicographical permutation.


## Complexity Analysis

Finding pivot = O(n)

Finding swap element = O(n)

Reversing suffix = O(n)

Total Time = O(n)

Space = O(1)


## Edge Cases

[1]
[1,1,1]
[3,2,1]
[1,2,3]
[1,5,1]


## Common Mistakes

1. Using reversed(nums) instead of nums.reverse().
2. Forgetting to return when pivot does not exist.
3. Searching for swap element in the wrong direction.
4. Sorting the suffix instead of reversing it.

## 30-Second Revision Notes

1. Find pivot from right.
2. Find rightmost greater element.
3. Swap.
4. Reverse suffix.
5. O(n) time, O(1) space.


```
class Solution:
    def nextPermutation(self, nums):
        n = len(nums)
        pivot = -1

        for i in range(n - 2, -1, -1):
            if nums[i] < nums[i + 1]:
                pivot = i
                break

        if pivot == -1:
            nums.reverse()
            return

        for i in range(n - 1, pivot, -1):
            if nums[i] > nums[pivot]:
                nums[i], nums[pivot] = nums[pivot], nums[i]
                break

        left, right = pivot + 1, n - 1
        while left < right:
            nums[left], nums[right] = nums[right], nums[left]
            left += 1
            right -= 1
```
