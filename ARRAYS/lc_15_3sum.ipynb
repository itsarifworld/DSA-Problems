# LeetCode 15 — 3Sum

## Problem Statement

Given an integer array `nums`, return all unique triplets:

```
[a, b, c]
```

such that:

```
a + b + c = 0
```

The solution set must not contain duplicate triplets.

Example:

```
Input:
nums = [-1,0,1,2,-1,-4]

Output:
[[-1,-1,2],[-1,0,1]]
```


# Key Insight

We need three numbers whose sum equals zero.

A brute force solution checks every possible triplet.

However, because order does not matter and duplicates are not allowed, sorting the array allows us to use the powerful Two Pointer technique.


# Solution 1: Triple Nested Loops (Brute Force)

## Idea

Try every possible combination of three numbers.

## Approach

1. Choose first number.
2. Choose second number.
3. Choose third number.
4. Check if sum equals zero.
5. Store unique triplets.


````
def threeSum(nums):
    result = set()
    n = len(nums)

    for i in range(n):
        for j in range(i + 1, n):
            for k in range(j + 1, n):

                if nums[i] + nums[j] + nums[k] == 0:
                    result.add(tuple(sorted(
                        [nums[i], nums[j], nums[k]]
                    )))

    return [list(x) for x in result]
````

## Complexity Analysis

### Time Complexity

Three nested loops:

```
n × n × n
```

Time Complexity:

**O(n³)**

### Space Complexity

Result storage ignored:

**O(1)**

Including output:

**O(k)**


# Solution 2: Sorting + Hash Set

## Idea

Fix one number.

Convert remaining problem into Two Sum.

Use a hash set to find the third number.

## Approach

1. Sort array.
2. Fix nums[i].
3. Search remaining elements using a hash set.
4. Avoid duplicates.


````
def threeSum(nums):
    nums.sort()
    n = len(nums)

    result = []

    for i in range(n):

        if i > 0 and nums[i] == nums[i-1]:
            continue

        seen = set()

        for j in range(i + 1, n):

            third = -(nums[i] + nums[j])

            if third in seen:
                result.append([nums[i], third, nums[j]])

                while j + 1 < n and nums[j] == nums[j + 1]:
                    j += 1

            seen.add(nums[j])

    return result
````

## Complexity Analysis

### Sorting

O(n log n)

### Outer Loop

O(n)

### Inner Loop

O(n)

Total:

O(n²)

### Space Complexity

Hash set can store up to n elements.

O(n)


# Solution 3: Sorting + Two Pointers (Optimal)

## Core Idea

After sorting:

For each element:

```
nums[i]
```

find two numbers whose sum equals:

```
-target = nums[i]
```

using two pointers.

This eliminates the need for a hash table.


## Detailed Approach

### Step 1: Sort Array

Example:

```
[-1,0,1,2,-1,-4]

becomes

[-4,-1,-1,0,1,2]
```

### Step 2: Fix First Element

Choose:

```
nums[i]
```

### Step 3: Use Two Pointers

```
left = i + 1
right = n - 1
```

Compute:

```
total = nums[i] + nums[left] + nums[right]
```

### Step 4: Move Pointers

If total < 0:

```
left += 1
```

Need larger sum.

If total > 0:

```
right -= 1
```

Need smaller sum.

If total == 0:

Store triplet.

Move both pointers while skipping duplicates.


````
def threeSum(nums):
    nums.sort()
    result = []

    n = len(nums)

    for i in range(n):

        if i > 0 and nums[i] == nums[i - 1]:
            continue

        left = i + 1
        right = n - 1

        while left < right:

            total = nums[i] + nums[left] + nums[right]

            if total < 0:
                left += 1

            elif total > 0:
                right -= 1

            else:

                result.append(
                    [nums[i], nums[left], nums[right]]
                )

                left += 1
                right -= 1

                while left < right and nums[left] == nums[left - 1]:
                    left += 1

                while left < right and nums[right] == nums[right + 1]:
                    right -= 1

    return result
````

# Why Sorting Helps

Without sorting:

- No direction information.
- Cannot decide which pointer to move.

After sorting:

If sum is too small:

Move left pointer right.

If sum is too large:

Move right pointer left.

This reduces the search from O(n³) to O(n²).


# Detailed Complexity Calculation (Optimal Solution)

Let n be the length of the array.

### Sorting

Python sort:

O(n log n)

### Outer Loop

Runs at most n times.

O(n)

### Inner Two Pointer Scan

For each i:

left and right move only forward/backward.

Across one iteration:

O(n)

Therefore:

Outer Loop × Inner Scan

= O(n × n)

= O(n²)

### Total Time Complexity

O(n log n) + O(n²)

Dominant term:

O(n²)

### Space Complexity

Ignoring output array:

Only pointers and variables are used.

O(1)

Including result storage:

O(k)


# Duplicate Handling

This is the most important part of 3Sum.

## Skip Duplicate First Elements

```python
if i > 0 and nums[i] == nums[i-1]:
    continue
```

## Skip Duplicate Left Values

```python
while left < right and nums[left] == nums[left-1]:
    left += 1
```

## Skip Duplicate Right Values

```python
while left < right and nums[right] == nums[right+1]:
    right -= 1
```

Without these checks, duplicate triplets appear in the answer.


# Interview Summary

| Solution | Time | Space |

| Sorting + Hash Set | O(n²) | O(n) |
| Sorting + Two Pointers | O(n²) | O(1) |

## Recommended Solution

Sorting + Two Pointers

Reasons:

- Most common interview solution.
- Optimal space complexity.
- Easy to extend to 4Sum and KSum.
- Demonstrates understanding of sorting and pointer techniques.
