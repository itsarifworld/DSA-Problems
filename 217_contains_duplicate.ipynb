# 217. Contains Duplicate

Given an integer array `nums`, return `True` if any value appears at least twice in the array, and return `False` if every element is distinct.

## Examples

```text
Input: nums = [1,2,3,1]
Output: True

Input: nums = [1,2,3,4]
Output: False
```

## Approach

Use a hash set to track numbers seen so far. If the current number already exists in the set, a duplicate has been found. Otherwise, add it and continue.

- Time complexity: `O(n)`
- Space complexity: `O(n)`

````python
from typing import List


class Solution:
    def containsDuplicate(self, nums: List[int]) -> bool:
        seen = set()

        for num in nums:
            if num in seen:
                return True
            seen.add(num)

        return False
````

````python
solution = Solution()

test_cases = [
    ([1, 2, 3, 1], True),
    ([1, 2, 3, 4], False),
    ([1, 1, 1, 3, 3, 4, 3, 2, 4, 2], True),
    ([], False),
]

for nums, expected in test_cases:
    result = solution.containsDuplicate(nums)
    assert result == expected, f"nums={nums}, expected={expected}, got={result}"

print("All test cases passed.")
````
