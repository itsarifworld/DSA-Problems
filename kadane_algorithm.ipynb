# Kadane's Algorithm

## Problem Statement
Given an integer array, find the contiguous subarray with the maximum sum.


```python
arr = [-2,-3,4,-1,-2,1,5,-3]
n = len(arr)

maxm = 0

for i in range(n):
    sum = 0
    for j in range(i, n):
        sum += arr[j]
        maxm = max(sum, maxm)

print(maxm)
```

## Optimized Approach (Kadane's Algorithm)

- Traverse the array
- Keep track of current sum
- Reset sum if it becomes negative
- Maintain maximum sum


```python
# Kadane's Algorithm (O(n))
arr = [-2,-3,4,-1,-2,1,5,-3]

max_sum = arr[0]
current_sum = 0

for num in arr:
    current_sum += num
    max_sum = max(max_sum, current_sum)
    if current_sum < 0:
        current_sum = 0

print(max_sum)
```
