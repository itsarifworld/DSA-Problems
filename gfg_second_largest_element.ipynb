
# Second Largest Element (GFG)

## Problem Statement
Given an array of positive integers `arr[]`, return the second largest element from the array.  
If the second largest element doesn't exist then return `-1`.

**Note:** The second largest element should not be equal to the largest element.

### Examples

- Input: `[12, 35, 1, 10, 34, 1]` → Output: `34`
- Input: `[10, 5, 10]` → Output: `5`
- Input: `[10, 10, 10]` → Output: `-1`

### Constraints
- 2 ≤ arr.size() ≤ 10^5  
- 1 ≤ arr[i] ≤ 10^5  



## Approach 1: Brute Force (Sorting)

Sort the array and traverse from the end to find the first element smaller than the maximum.

### Time Complexity: O(n log n)  
### Space Complexity: O(1)


```python

def second_largest_sort(arr):
    arr.sort()
    largest = arr[-1]

    for i in range(len(arr)-2, -1, -1):
        if arr[i] != largest:
            return arr[i]
    return -1
```


## Approach 2: Better (Two Pass)

1. Find the largest element  
2. Traverse again to find the largest element not equal to max  

### Time Complexity: O(n)  
### Space Complexity: O(1)


```python

def second_largest_two_pass(arr):
    largest = max(arr)
    second = -1

    for num in arr:
        if num != largest:
            second = max(second, num)

    return second
```


## Approach 3: Optimal (Single Pass)

Track largest and second largest in one traversal.

### Time Complexity: O(n)  
### Space Complexity: O(1)


```python

def second_largest_optimal(arr):
    if len(arr) < 2:
        return -1

    largest = float('-inf')
    second = float('-inf')

    for num in arr:
        if num > largest:
            second = largest
            largest = num
        elif num < largest and num > second:
            second = num

    return second if second != float('-inf') else -1
```

## Test Cases

```python

arr1 = [12, 35, 1, 10, 34, 1]
arr2 = [10, 5, 10]
arr3 = [10, 10, 10]

print(second_largest_optimal(arr1))  # 34
print(second_largest_optimal(arr2))  # 5
print(second_largest_optimal(arr3))  # -1
```
