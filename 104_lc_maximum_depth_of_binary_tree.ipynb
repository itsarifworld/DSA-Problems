# LeetCode 104 - Maximum Depth of Binary Tree

**Difficulty:** Easy  
**Topics:** Binary Tree, DFS, BFS, Recursion  
**Pattern:** Compute the height of a tree by exploring child subtrees.

## Problem Statement

Given the root of a binary tree, return its maximum depth.

The maximum depth is the number of nodes on the longest path from the root node down to a leaf node.

### Examples

```text
root = [3, 9, 20, null, null, 15, 7]
Output: 3
```

```text
root = [1, null, 2]
Output: 2
```

```text
root = []
Output: 0
```

## Core Idea

Depth is a recursive property:

```text
depth(root) = 1 + max(depth(root.left), depth(root.right))
```

The base case is an empty tree. If `root` is `None`, the depth is `0`.

This gives a direct recursive DFS solution. The same problem can also be solved iteratively with BFS by counting levels, or with an explicit DFS stack.

## Tree Node Setup

LeetCode provides the `TreeNode` class. The notebook defines it locally so the examples and tests can run outside LeetCode.

````python
from collections import deque
from typing import Optional


class TreeNode:
    def __init__(self, val: int = 0, left: Optional["TreeNode"] = None, right: Optional["TreeNode"] = None):
        self.val = val
        self.left = left
        self.right = right
````

## Approach 1: Recursive DFS

### Steps

1. If the current node is `None`, return `0`.
2. Recursively compute the maximum depth of the left subtree.
3. Recursively compute the maximum depth of the right subtree.
4. Return `1 + max(left_depth, right_depth)`.

### Why This Works

A tree's maximum depth is one node for the current root plus the deeper of its two child subtrees.

### Complexity

- Time: `O(n)` because every node is visited once.
- Space: `O(h)` recursion stack, where `h` is the tree height. Worst case is `O(n)` for a skewed tree.

````python
class RecursiveSolution:
    def maxDepth(self, root: Optional[TreeNode]) -> int:
        if root is None:
            return 0

        left_depth = self.maxDepth(root.left)
        right_depth = self.maxDepth(root.right)
        return 1 + max(left_depth, right_depth)
````

## Approach 2: Iterative BFS by Levels

### Idea

Breadth-first search visits the tree level by level. Each completed level increases the depth by `1`.

### Steps

1. If the root is empty, return `0`.
2. Put the root in a queue.
3. For each level, process exactly the nodes currently in the queue.
4. Add each processed node's children to the queue.
5. Increment the depth after finishing each level.

### Complexity

- Time: `O(n)`
- Space: `O(w)`, where `w` is the maximum width of the tree.

````python
class BFSSolution:
    def maxDepth(self, root: Optional[TreeNode]) -> int:
        if root is None:
            return 0

        queue = deque([root])
        depth = 0

        while queue:
            for _ in range(len(queue)):
                node = queue.popleft()
                if node.left:
                    queue.append(node.left)
                if node.right:
                    queue.append(node.right)
            depth += 1

        return depth
````

## Approach 3: Iterative DFS with Depth Tracking

### Idea

Use a stack of `(node, depth)` pairs. When a node is popped, update the best depth seen so far and push its children with `depth + 1`.

This avoids recursion while keeping the same depth-first behavior.

### Complexity

- Time: `O(n)`
- Space: `O(h)` on balanced trees and `O(n)` in the worst case.

````python
class Solution:
    def maxDepth(self, root: Optional[TreeNode]) -> int:
        if root is None:
            return 0

        max_depth = 0
        stack = [(root, 1)]

        while stack:
            node, depth = stack.pop()
            max_depth = max(max_depth, depth)

            if node.left:
                stack.append((node.left, depth + 1))
            if node.right:
                stack.append((node.right, depth + 1))

        return max_depth
````

## Recursive Walkthrough

For this tree:

```text
        3
       / \
      9  20
         / \
        15  7
```

- Node `9` has no children, so its depth is `1`.
- Nodes `15` and `7` also have depth `1`.
- Node `20` has depth `1 + max(1, 1) = 2`.
- Root `3` has depth `1 + max(1, 2) = 3`.

The answer is `3`.

## Test Helper

The helper below builds a binary tree from a level-order list, using `None` for missing nodes.

````python
def build_tree(values: list[Optional[int]]) -> Optional[TreeNode]:
    if not values:
        return None

    nodes = [None if value is None else TreeNode(value) for value in values]
    children = nodes[::-1]
    root = children.pop()

    for node in nodes:
        if node is None:
            continue
        if children:
            node.left = children.pop()
        if children:
            node.right = children.pop()

    return root
````

````python
solver = Solution()

assert solver.maxDepth(build_tree([3, 9, 20, None, None, 15, 7])) == 3
assert solver.maxDepth(build_tree([1, None, 2])) == 2
assert solver.maxDepth(build_tree([])) == 0
assert solver.maxDepth(build_tree([1, 2, 3, 4, 5])) == 3
assert solver.maxDepth(build_tree([1, 2, None, 3, None, 4])) == 4

print("All test cases passed.")
````

## Takeaways

- Recursive DFS is the cleanest solution for maximum tree depth.
- BFS is useful when thinking in terms of levels.
- Iterative DFS avoids recursion limits and makes the tracked depth explicit.
- Empty subtrees contribute depth `0`; non-empty nodes contribute `1` plus the deeper child depth.
