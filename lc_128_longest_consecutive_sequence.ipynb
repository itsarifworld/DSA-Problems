
# 🧠 Longest Consecutive Sequence (LeetCode 128)

## Problem Statement
Given an unsorted array of integers `nums`, return the length of the longest consecutive elements sequence.

You must write an algorithm that runs in **O(n)** time.

---

### Example 1:
Input: nums = [100,4,200,1,3,2]  
Output: 4  
Explanation: The longest sequence is [1,2,3,4]

### Example 2:
Input: nums = [0,3,7,2,5,8,4,6,0,1]  
Output: 9

### Example 3:
Input: nums = [1,0,1,2]  
Output: 3



## 🔴 Brute Force Approach

### Idea:
- For each number, try to find next consecutive numbers using linear search.
- Keep counting until sequence breaks.

### Time Complexity:
- O(n³) (very inefficient)

### Intuition:
For every element → check if next exists → keep checking


```

def longestConsecutive_brute(nums):
    longest = 0

    for num in nums:
        current = num
        length = 1

        while current + 1 in nums:
            current += 1
            length += 1

        longest = max(longest, length)

    return longest

# Test
print(longestConsecutive_brute([100,4,200,1,3,2]))
```


## 🟡 Better Approach (Sorting)

### Idea:
- Sort the array
- Traverse and count consecutive elements

### Time Complexity:
- O(n log n)

### Key Insight:
Sorting helps bring consecutive numbers together


```

def longestConsecutive_better(nums):
    if not nums:
        return 0

    nums = sorted(set(nums))  # remove duplicates + sort
    longest = 1
    length = 1

    for i in range(1, len(nums)):
        if nums[i] == nums[i-1] + 1:
            length += 1
        else:
            length = 1

        longest = max(longest, length)

    return longest

# Test
print(longestConsecutive_better([100,4,200,1,3,2]))
```


## 🟢 Optimal Approach (HashSet)

### Idea:
- Use a set for O(1) lookups
- Only start counting when number is the **start of a sequence**
  (i.e., num - 1 not present)

### Why this works:
Avoids recomputing sequences multiple times

### Time Complexity:
- O(n)

### Key Insight:
Only start when it's the **first element of sequence**


```

def longestConsecutive_optimal(nums):
    num_set = set(nums)
    longest = 0

    for num in num_set:
        if num - 1 not in num_set:
            current = num
            length = 1

            while current + 1 in num_set:
                current += 1
                length += 1

            longest = max(longest, length)

    return longest

# Test
print(longestConsecutive_optimal([100,4,200,1,3,2]))
```


## ⚡ Final Comparison

| Approach | Time Complexity | Space | Notes |
|---------|---------------|------|------|
| Brute   | O(n³)         | O(1) | Too slow |
| Better  | O(n log n)    | O(n) | Uses sorting |
| Optimal | O(n)          | O(n) | Best solution |

---

## 🚀 Key Takeaway
The trick is not just hashing — it's **avoiding unnecessary sequence checks**.
