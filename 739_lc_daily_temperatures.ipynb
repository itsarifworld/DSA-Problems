# LeetCode 739 - Daily Temperatures

**Difficulty:** Medium  
**Topics:** Array, Stack, Monotonic Stack  
**Pattern:** Keep unresolved indices in a stack until a warmer temperature appears.

## Problem Statement

You are given an array `temperatures`, where `temperatures[i]` is the temperature on day `i`.

For every day, return how many days must pass until a warmer temperature occurs. If there is no future day with a warmer temperature, return `0` for that day.

### Examples

```text
temperatures = [73, 74, 75, 71, 69, 72, 76, 73]
Output:       [1, 1, 4, 2, 1, 1, 0, 0]
```

```text
temperatures = [30, 40, 50, 60]
Output:       [1, 1, 1, 0]
```

```text
temperatures = [30, 60, 90]
Output:       [1, 1, 0]
```

## Core Idea

For each day, we need the next day to the right with a strictly higher temperature.

This is a classic **next greater element** problem. The efficient way to solve it is to keep a stack of days whose answer has not been found yet.

When today is warmer than the day at the top of the stack, today becomes the answer for that earlier day.

## Approach 1: Brute Force

### Steps

1. For each day `i`, scan every later day `j`.
2. Stop at the first `j` where `temperatures[j] > temperatures[i]`.
3. The answer for day `i` is `j - i`.
4. If no warmer day exists, keep the answer as `0`.

### Complexity

- Time: `O(n^2)` in the worst case.
- Space: `O(1)` extra space, not counting the output array.

This is correct but too slow for large inputs.

````python
from typing import List


class BruteForceSolution:
    def dailyTemperatures(self, temperatures: List[int]) -> List[int]:
        n = len(temperatures)
        answer = [0] * n

        for i in range(n):
            for j in range(i + 1, n):
                if temperatures[j] > temperatures[i]:
                    answer[i] = j - i
                    break

        return answer
````

## Approach 2: Monotonic Decreasing Stack

### Stack Meaning

The stack stores indices of days that are still waiting for a warmer future day.

The temperatures at those indices are kept in decreasing order from bottom to top. This is why it is called a **monotonic decreasing stack**.

### Steps

1. Create an output array filled with `0`.
2. Traverse the temperatures from left to right.
3. While the stack is not empty and today's temperature is warmer than the temperature at the stack's top index:
   - Pop that previous index.
   - The wait time is `current_index - previous_index`.
4. Push the current index onto the stack.

### Why This Works

Each day waits on the stack until the first warmer day appears. Because we process days from left to right, the first warmer day that pops an index is automatically the nearest warmer future day.

Each index is pushed once and popped at most once, so the total work is linear.

### Complexity

- Time: `O(n)`
- Space: `O(n)` for the stack and output array.

````python
class Solution:
    def dailyTemperatures(self, temperatures: List[int]) -> List[int]:
        answer = [0] * len(temperatures)
        stack = []  # indices waiting for a warmer temperature

        for current_day, current_temp in enumerate(temperatures):
            while stack and current_temp > temperatures[stack[-1]]:
                previous_day = stack.pop()
                answer[previous_day] = current_day - previous_day

            stack.append(current_day)

        return answer
````

## Walkthrough

For `temperatures = [73, 74, 75, 71, 69, 72, 76, 73]`:

- Day `0` has `73`. Stack: `[0]`.
- Day `1` has `74`, which is warmer than day `0`. Answer for day `0` is `1 - 0 = 1`.
- Day `2` has `75`, which is warmer than day `1`. Answer for day `1` is `2 - 1 = 1`.
- Days `3` and `4` are cooler, so they wait on the stack.
- Day `5` has `72`, which resolves day `4` and day `3`.
- Day `6` has `76`, which resolves day `5` and day `2`.
- Day `7` has `73`, but no later warmer day exists.

Final answer: `[1, 1, 4, 2, 1, 1, 0, 0]`.

````python
solver = Solution()

assert solver.dailyTemperatures([73, 74, 75, 71, 69, 72, 76, 73]) == [1, 1, 4, 2, 1, 1, 0, 0]
assert solver.dailyTemperatures([30, 40, 50, 60]) == [1, 1, 1, 0]
assert solver.dailyTemperatures([30, 60, 90]) == [1, 1, 0]
assert solver.dailyTemperatures([90, 80, 70]) == [0, 0, 0]
assert solver.dailyTemperatures([70, 70, 71]) == [2, 1, 0]

print("All test cases passed.")
````

## Takeaways

- This is a next-greater-element problem.
- A monotonic stack is useful when each element needs to find the next greater or smaller element on one side.
- Store indices instead of values when the answer depends on distance or position.
- The stack approach is linear because every index is pushed once and popped at most once.
