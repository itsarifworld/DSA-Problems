# 🧩 56. Merge Intervals (LeetCode)

## 📌 Problem Statement
Given an array of intervals where each interval is `[start, end]`, merge all overlapping intervals and return non-overlapping intervals.

### Constraints
- 1 <= intervals.length <= 10^4
- 0 <= start <= end <= 10^4

---

## 🔍 Examples

### Example 1
```
Input: [[1,3],[2,6],[8,10],[15,18]]
Output: [[1,6],[8,10],[15,18]]
```

### Example 2
```
Input: [[1,4],[4,5]]
Output: [[1,5]]
```

---

## 🧠 Intuition

Key idea:

- Overlap condition → `prev_end >= curr_start`
- If overlap → merge
- Else → new interval

👉 Sorting ensures overlapping intervals come together.

---

## 🎥 Step-by-Step Visualization

```
Input: [[1,3],[2,6],[8,10],[15,18]]

Sorted:
[[1,3],[2,6],[8,10],[15,18]]

Step 1: [1,3]
Step 2: [2,6] → overlap → [1,6]
Step 3: [8,10] → no overlap
Step 4: [15,18] → no overlap

Final:
[[1,6],[8,10],[15,18]]
```


````python
from typing import List

class Solution:
    def merge(self, intervals: List[List[int]]) -> List[List[int]]:
        new = []
        n = len(intervals)

        # Sorting
        intervals.sort()

        # Merging
        for i in range(n):
            if not new or intervals[i][0] > new[-1][1]:
                new.append(intervals[i])
            else:
                new[-1][1] = max(new[-1][1], intervals[i][1])

        return new
````

````python
# Dry Run Simulation
intervals = [[1,3],[2,6],[8,10],[15,18]]
intervals.sort()

merged = []

for interval in intervals:
    print("Current:", interval)
    if not merged or interval[0] > merged[-1][1]:
        merged.append(interval)
    else:
        merged[-1][1] = max(merged[-1][1], interval[1])
    print("Merged:", merged, "\n")

print("Final:", merged)
````

## ⏱️ Complexity Analysis

### Time Complexity
- Sorting → O(n log n)
- Traversal → O(n)
- **Total → O(n log n)**

### Space Complexity
- Output list → O(n)

---

## ⚠️ Edge Cases

- Single interval
- Fully overlapping intervals
- Touching intervals ([1,4],[4,5])
- Reverse input (unsorted)

---

## 🚀 Final Takeaway

👉 Pattern:
```
Sort → Iterate → Merge
```

This is one of the most important interval patterns in interviews.
