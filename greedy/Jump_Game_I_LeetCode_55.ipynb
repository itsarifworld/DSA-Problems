# Jump Game I (LeetCode 55)

## Problem
Given an integer array `nums`, where each element represents the maximum jump length from that position, determine whether you can reach the last index.

### Example
```text
Input: nums = [2,3,1,1,4]
Output: True
```

---

## Idea (Greedy)

Keep track of the **farthest index** that can be reached.

- If the current index is greater than the farthest reachable index, we cannot move further.
- Otherwise, continuously update the farthest reachable index.
- If the farthest reachable index reaches or crosses the last index, return `True`.

**Time Complexity:** `O(n)`

**Space Complexity:** `O(1)`


````
from typing import List

class Solution:
    def canJump(self, nums: List[int]) -> bool:
        farthest = 0

        for i in range(len(nums)):
            if i > farthest:
                return False

            farthest = max(farthest, i + nums[i])

            if farthest >= len(nums) - 1:
                return True

        return True
````

## Dry Run

For `nums = [2,3,1,1,4]`

| Index | Max Jump | Farthest |
|------:|---------:|---------:|
| 0 | 2 | 2 |
| 1 | 3 | 4 |
| Reach last index | ✅ | True |

---

## Key Takeaway

Instead of checking every possible jump, always remember the **maximum reachable index**. If you ever reach an index beyond it, the answer is `False`.
