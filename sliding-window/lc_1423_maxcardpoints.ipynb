# LeetCode 1423: Maximum Points You Can Obtain from Cards

## Intuition
- We can only take cards from the left end or right end.
- Start by taking all `k` cards from the left.
- Gradually replace cards taken from the left with cards from the right.
- Track the maximum score obtained.


## Solution Code

```python
from typing import List

class Solution:
    def maxScore(self, cardPoints: List[int], k: int) -> int:
        Lsum = 0
        Rsum = 0
        MaxSum = 0

        n = len(cardPoints)

        for i in range(0, k):
            Lsum = Lsum + cardPoints[i]

        MaxSum = Lsum
        right = n - 1

        for i in range(k - 1, -1, -1):
            Lsum = Lsum - cardPoints[i]
            Rsum = Rsum + cardPoints[right]
            right = right - 1
            MaxSum = max(MaxSum, Lsum + Rsum)

        return MaxSum
```

## Complexity Analysis

- Time Complexity: **O(k)**
- Space Complexity: **O(1)**
