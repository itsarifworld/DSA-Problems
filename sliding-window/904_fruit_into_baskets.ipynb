# LeetCode 904 — Fruit Into Baskets

## Problem Statement

You are given an integer array `fruits` where `fruits[i]` is the type of fruit produced by the `i-th` tree.

You want to collect as much fruit as possible. However:

- You only have **two baskets**.
- Each basket can hold **only one type of fruit**, but any quantity of that type.
- You must start at some tree and move to the right.
- Once you start, you must pick exactly one fruit from every tree until you encounter a third fruit type.

Return the maximum number of fruits you can collect.


# Solution 1 — Brute Force

For every starting index, extend the subarray until more than two fruit types appear.

### Time Complexity
`O(n²)`

### Space Complexity
`O(1)` to `O(3)` (effectively constant)


````
from typing import List

class Solution:
    def totalFruit(self, fruits: List[int]) -> int:
        n = len(fruits)
        ans = 0

        for i in range(n):
            freq = {}

            for j in range(i, n):
                freq[fruits[j]] = freq.get(fruits[j], 0) + 1

                if len(freq) > 2:
                    break

                ans = max(ans, j - i + 1)

        return ans
````

# Solution 2 — Optimal Sliding Window

## Key Observation

We need the **longest contiguous subarray containing at most 2 distinct values**.

This is a classic sliding window problem.

### Window Rules

- Expand the right pointer.
- Store fruit frequencies in a dictionary.
- If distinct fruit types become greater than 2:
  - Shrink from the left.
  - Decrease frequency.
  - Remove fruit type when frequency becomes 0.
- Track the maximum valid window length.

### Why It Works

At any moment, the window represents a valid segment containing at most two fruit types.

Every element enters and leaves the window at most once, giving linear complexity.


````
from typing import List

class Solution:
    def totalFruit(self, fruits: List[int]) -> int:
        l = 0
        maxlen = 0
        freq = {}

        for r in range(len(fruits)):
            freq[fruits[r]] = freq.get(fruits[r], 0) + 1

            while len(freq) > 2:
                freq[fruits[l]] -= 1

                if freq[fruits[l]] == 0:
                    freq.pop(fruits[l])

                l += 1

            maxlen = max(maxlen, r - l + 1)

        return maxlen
````
