# LeetCode 3: Longest Substring Without Repeating Characters

```
class Solution:
    def lengthOfLongestSubstring(self, s: str) -> int:
        # HASHMAP TO STORE NUMBER OF EACH letter
        freq = {}
        n = len(s)
        left = 0
        best = 0

        for i in range(n):
            freq[s[i]] = freq.get(s[i], 0) + 1

            while freq[s[i]] > 1:
                freq[s[left]] -= 1
                left += 1

            best = max(best, i - left + 1)

        return best
```
