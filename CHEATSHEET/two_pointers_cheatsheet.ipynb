# Two Pointers Technique (Opposite Ends)

A powerful pattern used in arrays and strings where two pointers start from opposite ends and move toward each other based on conditions.

## Importance
- Reduces time complexity from O(n^2) to O(n) in many problems
- Avoids nested loops
- Clean and intuitive once understood
- Common in interviews and competitive programming

## Where It Is Used
- Sorted arrays (pair sum problems)
- Palindrome checking
- Removing duplicates
- Container with most water
- Trapping rainwater (advanced variant)

## Step-by-Step Intuition
1. Initialize two pointers:
   - `left` at the beginning
   - `right` at the end

2. Loop while `left < right`

3. Perform logic based on the problem

4. Move pointers:
   - If condition satisfied → move `left`
   - Otherwise → move `right`

5. Continue until pointers meet

```python
# Template Code
def fn(arr):
    left = ans = 0
    right = len(arr) - 1

    while left < right:
        # do some logic here with left and right
        if CONDITION:
            left += 1
        else:
            right -= 1
    
    return ans
```

## Example 1: Check if Array is Palindrome
Compare elements from both ends moving inward.

```python
def is_palindrome(arr):
    left, right = 0, len(arr) - 1
    
    while left < right:
        if arr[left] != arr[right]:
            return False
        left += 1
        right -= 1
    
    return True

print(is_palindrome([1, 2, 3, 2, 1]))  # True
```

## Example 2: Two Sum in Sorted Array
Find if there exists a pair that sums to target.

```python
def two_sum_sorted(arr, target):
    left, right = 0, len(arr) - 1

    while left < right:
        curr_sum = arr[left] + arr[right]

        if curr_sum == target:
            return (left, right)
        elif curr_sum < target:
            left += 1
        else:
            right -= 1

    return -1

print(two_sum_sorted([1,2,3,4,6], 6))  # (1, 3)
```

## Example 3: Container With Most Water
Maximize area using two pointers.

```python
def max_area(height):
    left, right = 0, len(height) - 1
    max_water = 0

    while left < right:
        width = right - left
        h = min(height[left], height[right])
        max_water = max(max_water, width * h)

        if height[left] < height[right]:
            left += 1
        else:
            right -= 1

    return max_water

print(max_area([1,8,6,2,5,4,8,3,7]))  # 49
```
