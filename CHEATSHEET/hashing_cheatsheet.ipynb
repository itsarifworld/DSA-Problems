# 📓 Hashing Cheatsheet (Python)
Clean, structured, and exam-ready.

## 🟦 Problem Statement
Efficiently store and retrieve data using hashing for **O(1)** average operations.

## 🟦 Example: Frequency Count

```python
arr = [1, 2, 2, 3, 3, 3]

freq = {}
for num in arr:
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1

print(freq)
```

## 🟦 Intuition
- Hashing = Instant lookup
- Use dictionary for frequency
- Use set for duplicates
- Think: "Have I seen this before?"


## 🟦 Step-by-Step Working

```python
arr = [4, 5, 4, 6, 5]

freq = {}

for num in arr:
    if num in freq:
        freq[num] += 1
    else:
        freq[num] = 1

print(freq)
```

## 🟦 Pythonic Version

```python
from collections import Counter

arr = [4, 5, 4, 6, 5]
freq = Counter(arr)

print(freq)
```

## 🟩 Problem 1: Two Sum

```python
def two_sum(nums, target):
    hashmap = {}
    for i, num in enumerate(nums):
        diff = target - num
        if diff in hashmap:
            return [hashmap[diff], i]
        hashmap[num] = i

print(two_sum([2,7,11,15], 9))
```

## 🟩 Problem 2: First Non-Repeating Character

```python
from collections import Counter

def first_unique(s):
    freq = Counter(s)
    for ch in s:
        if freq[ch] == 1:
            return ch
    return -1

print(first_unique("aabbcde"))
```

## 🟩 Problem 3: Contains Duplicate

```python
def contains_duplicate(nums):
    seen = set()
    for num in nums:
        if num in seen:
            return True
        seen.add(num)
    return False

print(contains_duplicate([1,2,3,1]))
```

## 🟩 Problem 4: Longest Subarray with Sum K

```python
def longest_subarray(nums, k):
    prefix_sum = 0
    hashmap = {}
    max_len = 0

    for i in range(len(nums)):
        prefix_sum += nums[i]

        if prefix_sum == k:
            max_len = i + 1

        if (prefix_sum - k) in hashmap:
            max_len = max(max_len, i - hashmap[prefix_sum - k])

        if prefix_sum not in hashmap:
            hashmap[prefix_sum] = i

    return max_len

print(longest_subarray([1,2,3,1,1,1], 3))
```

## 🟩 Problem 5: Group Anagrams

```python
from collections import defaultdict

def group_anagrams(strs):
    hashmap = defaultdict(list)
    for word in strs:
        key = "".join(sorted(word))
        hashmap[key].append(word)
    return list(hashmap.values())

print(group_anagrams(["eat","tea","tan","ate","nat","bat"]))
```
