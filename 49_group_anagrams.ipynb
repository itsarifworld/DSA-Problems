# 49. Group Anagrams

Given an array of strings `strs`, group the anagrams together. You can return the answer in any order.

An anagram is a word formed by rearranging the letters of another word using all original letters exactly once.

## Examples

```text
Input: strs = ["eat", "tea", "tan", "ate", "nat", "bat"]
Output: [["bat"], ["nat", "tan"], ["ate", "eat", "tea"]]

Input: strs = [""]
Output: [[""]]

Input: strs = ["a"]
Output: [["a"]]
```

## Approach

Use a hash map where each key represents the sorted character order of a word. All words with the same sorted key are anagrams, so they belong in the same group.

- Time complexity: `O(n * k log k)`, where `n` is the number of strings and `k` is the maximum string length
- Space complexity: `O(n * k)`

````python
from collections import defaultdict
from typing import List


class Solution:
    def groupAnagrams(self, strs: List[str]) -> List[List[str]]:
        groups = defaultdict(list)

        for word in strs:
            key = "".join(sorted(word))
            groups[key].append(word)

        return list(groups.values())
````

````python
def normalize(groups: List[List[str]]) -> List[List[str]]:
    return sorted([sorted(group) for group in groups])


solution = Solution()

test_cases = [
    (["eat", "tea", "tan", "ate", "nat", "bat"], [["bat"], ["nat", "tan"], ["ate", "eat", "tea"]]),
    ([""], [[""]]),
    (["a"], [["a"]]),
    (["abc", "bca", "cab", "xyz", "zyx"], [["abc", "bca", "cab"], ["xyz", "zyx"]]),
]

for strs, expected in test_cases:
    result = solution.groupAnagrams(strs)
    assert normalize(result) == normalize(expected), f"strs={strs}, expected={expected}, got={result}"

print("All test cases passed.")
````
