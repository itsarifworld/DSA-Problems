# Missing and Repeating Number — Complete Guide 🚀


## Problem Statement

Given an array of size **n** containing numbers from **1 to n**:
- One number is **missing**
- One number is **repeating**

### Example
```
Input: [4, 3, 6, 2, 1, 1]
Output: Missing = 5, Repeating = 1
```



## 🔍 Intuition

- Numbers should be exactly from **1 to n**
- If one repeats → one must be missing
- Total count stays same → we exploit this mathematically


## 1. Brute Force Approach (O(n²))


### Idea
Check each number from 1 to n and count occurrences.

### Pros
- Easy to understand

### Cons
- Very slow for large n


````

arr = [4, 3, 6, 2, 1, 1]
n = len(arr)

repeating = -1
missing = -1

for i in range(1, n+1):
    cnt = 0
    for j in range(n):
        if arr[j] == i:
            cnt += 1

    if cnt == 2:
        repeating = i
    elif cnt == 0:
        missing = i

print("Missing:", missing)
print("Repeating:", repeating)
````

👉 We scan the array again and again → that's why O(n²)

## 2. Hashing Approach (O(n))


### Idea
Use a frequency array to store counts.

### Pros
- Fast
- Simple logic

### Cons
- Extra space required


````

arr = [4, 3, 6, 2, 1, 1]
n = len(arr)

freq = [0]*(n+1)

for num in arr:
    freq[num] += 1

for i in range(1, n+1):
    if freq[i] == 2:
        repeating = i
    elif freq[i] == 0:
        missing = i

print("Missing:", missing)
print("Repeating:", repeating)
````

👉 Frequency array directly tells us what's missing and repeating

## 3. Mathematical Approach (Optimal)


### Key Formulas

- Sum = n(n+1)/2  
- Sum of squares = n(n+1)(2n+1)/6  

### Let:
- X = repeating  
- Y = missing  

Then:
- X - Y = difference of sums  
- X + Y = derived from square sums  

Solve → get X and Y


````

arr = [4, 3, 6, 2, 1, 1]
n = len(arr)

Sarr = sum(arr)
Sn = (n*(n+1))//2

S2arr = sum(i*i for i in arr)
S2n = (n*(n+1)*(2*n+1))//6

sub = Sarr - Sn          # X - Y
add = (S2arr - S2n)//sub  # X + Y

repeating = (sub + add)//2
missing = add - repeating

print("Missing:", missing)
print("Repeating:", repeating)
````

👉 Best approach for interviews (O(n), O(1) space)

## 4. XOR Approach (Advanced)


### Idea
- XOR all array elements and numbers 1..n
- Result = X ^ Y
- Separate using rightmost set bit

### Pros
- No overflow risk
- O(1) space

### Cons
- Hard to understand initially


````

arr = [4, 3, 6, 2, 1, 1]
n = len(arr)

xor = 0

for i in arr:
    xor ^= i

for i in range(1, n+1):
    xor ^= i

# rightmost set bit
bit = xor & -xor

x = 0
y = 0

for i in arr:
    if i & bit:
        x ^= i
    else:
        y ^= i

for i in range(1, n+1):
    if i & bit:
        x ^= i
    else:
        y ^= i

# determine which is repeating
if arr.count(x) == 2:
    repeating = x
    missing = y
else:
    repeating = y
    missing = x

print("Missing:", missing)
print("Repeating:", repeating)
````


## ⚠️ Edge Cases

- Array size = 2
- Repeating number at start/end
- Large numbers (prefer XOR or math)




## 🏁 Final Comparison

| Approach | Time | Space | Difficulty |
|--------|------|------|------------|
| Brute Force | O(n²) | O(1) | Easy |
| Hashing | O(n) | O(n) | Medium |
| Math | O(n) | O(1) | Medium |
| XOR | O(n) | O(1) | Hard |
